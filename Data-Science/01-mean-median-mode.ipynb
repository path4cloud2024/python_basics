## Central Tendency : Mean, Median and Mode
- A lot of operation can be doable with built in functions which comes with standard library in Python Installation

```python
# let say, we need to find the avg of a given list of marks
```

```python
# used sum function to get the 
# used len function to get the length
# then we divided it to get the avg = sum / total
marks = [93, 35, 78, 35, 85]
sum(marks) / len(marks)
```

# But better to do it with dedicated library or module if we have available
- will import the statistics (entire module) which comes with lot of functions
- if we just type statistics after import then put a dot(.) and hit TAB button, it will show all functions or methods associated with this

```python
import statistics
```

```python
statistics.mean(marks)
```

```python
statistics.median(marks) 
```

```python
statistics.mode(marks)
```
