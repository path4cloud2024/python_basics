# Decimal Library
- Decimal Type (not builtin, additional library which are predefined compatability with Python)
- sometime, we need to deal with monetary amount or some exact numbers especailly floating number. By default, Python stores the float number as binary and if we need to print some additinoal decimal values on the right side of dot in value, then it might give you some buggy number which would be an issue if we are dealing with monetary amounts.

```python
# Python can deal with float values but upto some certain decimal values
x = 15.5
y = 2.0
x + y
```

```python
# but what if we have multiple decimal digits, then it fails with representational error
x = 20345.3476327468
y = 213.00
x + y
```

### Python allocates 8 bytes to store float's values, so stores value can be only approximation
### If we have infinite number or dealing with some financial transaction then we might face a loss or loose some number after dots

```python
# example
x = 10.0
y = 3.0
x / y
```

### Here we can see last digit is rounded to 5 beacuse it is infinite 3 post dots. 
### Many applications require precise represented of numbers with decimal points

## To deal with this situation we have `decimal` library which we can import and tackle this situation

```python
from decimal import Decimal
```

```python
x = Decimal('10.0')
y = Decimal('3.0')
x / y # will print in decimal string representation
```

```python
# If we use print with same operation then it will print just printable representation
print(x / y)
```

### If amount need to put with some decimal points

```python
# without decimal 
amount = 121.20
amount
```

```python
# with f string
print(f'{amount:.20f}') # <--- here, it will give some representational error
```

```python
# with decimal
amount1 = Decimal('121.20')
print(f'{amount1:.20f}') # <--- here, no representation error
```

# some pros & cons for Decimal Library
- Uses special coding scheme to ensure pricise number
- But this scheme requires additional memory to store numbers
- Additional proceesing time to perform calculations
- But proovides precision required for monetary calculation

- Solve a Compound Interest problem with Decimal
- A person invests  1000 INR in a savings account with 5% interest. Assuming that the person leaves all interest on deposit in the account, calculate and display the account balance at the end of each year for 10 years. Use this formula for determining the amounts:
        a = p(1+r)n (n is the power of (1+r))
- Hint:
  - We can't perform mathematical operations between Decimal and Float (will throw an error)
  - We can perform mathematical operations between Decimal and Decimal OR Decimal and Integer
  - example below:


```python
Decimal(10.5) * 5.0
```

```python
Decimal(10.5) * Decimal(5.0)
```

```python
Decimal(10.5) * 5
```

```python
# SOLUTION
principal = Decimal('1000.00')
rate = Decimal('0.05')
for year in range(1,11):
    amount = principal * (1 + rate) ** year
    print(f'{year:>2}{amount:>10.2f}')
    ## here >2 and >10 are used, > is a field with a character which will align the value from right to the number of spaces mentioned
    ## In this case, the year will fit the 2 spaces from the right, and the amount will fit the 10 spaces from the right of the screen
```
