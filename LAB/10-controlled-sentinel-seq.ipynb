# controlled sequence

```python
money = 0
month = 0
spent_money = [24300, 33400, 41000, 23950, 18450]
```

```python
for expenses in spent_money:
    money += expenses
    month += 1
```

```python
avg_money = money / month
print(f'Average expenses in {month} months is {avg_money}')
```

# sentinel controlled

```python
money1 = 0
month1 = 0
spent_money1 = int(input("enter the amount you spent, -1 to exit: "))

# We don't have the ending iteration, so we use a while loop
while spent_money1 != -1 :
    money1 += spent_money1
    month1 += 1
    spent_money1 = int(input("enter the amount you spent, -1 to exit: "))

if month1 != 0:
    avg_money1 = money1 / month1
    print(f'Avergae expenses in {month1} month is {avg_money1:.2f}') # here :.2f after variable print the two decimal values at end.
else:
    print("you quit the calculation")
```
