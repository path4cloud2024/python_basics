## Random Library
- random.random() gives a value between 0 and 1

```python
import random
```

```python
random.random()
```

### In above example, random.random() produced some random number with multiple decimal points.
### What if we need to format that then we have following way:
- using `round()` function: e.g. round(random.random(), 2)   # 2 decimal places
- using `f-sting` : print(f'{random.random():.2f}') # 2 decimal places
- using `format` : print("{:.2f}".format(random.random()))

### `f-string` is the recommend and professional way

```python
# using round function
num = random.random()
print(round(num, 2))
```

```python
# using f-string
num = random.random()
print(f'{num:.2f}')
```

```python
# using format
num = random.random()
print('{:.2f}'.format(num))
```

### SEED Function
- It controls the randomness
- It makes random numbers repeatable (predictable)
- Use Case:
  - Testing Automation
  - Game Development, to recreate the same game scenario
  - Debugging: If a bug happens with random values, you can reproduce the issue exactly

- from this point, will introduce the seed function and from now onwards, will run the for loop 3 times, and will get each time
- new values, then will again run the seed function, we will see, then all 3 values will be repeated.

```python
random.seed(3)
```

```python
for new_num in range(5):
    print(random.randrange(1,3), end=' ')
```

```python
for new_num in range(5):
    print(random.randrange(1,3), end=' ')
```

```python
for new_num in range(5):
    print(random.randrange(1,3), end=' ')
```

- Repeating the same seed funciton with same value, now you can see, we are getting the same value from first run of for loop after seed functions above

- Here, random.seed(3) is a random generator and each time will give you a unique range of random numbers
- Using some internal calculation. Until you call it a generator again, it will keep on giving random range, but once you recall it, again the same sequence of random range will be followed, which was already generated

# Two Types of Randomness in Python
- Pseudorandom → random module
- Cryptographically secure → secrets module
- Core Difference



```python
import random
otp = random.randint(1000, 9999)
print(otp)
```

```python
import secrets
otp1 = secrets.randbelow(9000) + 1000
print(otp1)
```

## PRO TIP
- Secrets is more secure because it uses OS-level entropy (mouse movement, CPU timing, hardware noise), making outputs unpredictable, while random is deterministic and can be predicted if the seed or state is known.
