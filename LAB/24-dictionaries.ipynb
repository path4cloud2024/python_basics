# Dictionaries

- Dictionaries for collections of key-value pairs, each key maps to a specific value and is denoted as `{ }` in Python
- Keys must be `immutable` and `unique`; no duplicate keys are allowed
- Dictionaries were an unordered set of collections till before Python 3.7, but Python 3.7 and later are ordered collections.
  - They preserve insertion order
  - Items stay in the order you added them
- Values can be of any mutable or immutable type
- Multiple keys can have the same value
- Dictionary doesn't support indexing/slicing
- Create, initialize, and refer to dictionary elements
- Iterate over a dictionary's keys, values, and key-value pairs
- Add, remove, and update a dictionary's key-value pairs
- Use a dictionary with comparison operators
- Functional Programming - Dictionary Comprehensions

```python
# Example
student = {
    "name": "Amit",
    "age": 22,
    "course": "Python"
}
```

```python
# Accessing Values
print(student["name"])   # Output: Amit
print(student["age"])    # Output: 22
```

## Key Features of Dictionary

- Keys must be unique
- Keys can be different types (mostly immutable)
- Values can be anything
- Common Operations
  - Add / Update
  - Delete
  - Loop through dictionary

```python
#unique key
data = {"a": 1, "a": 2}
print(data)   # Output: {'a': 2} will get only last update key:value
```

```python
# different key
data = {
    1: "one",
    "name": "Amit",
    (1,2): "tuple key"
}
print(data)
```

```python
# value can be any data type
data = {
    "list": [1,2,3],
    "dict": {"a": 10}
}
# access the value of the key "list."
print(data["list"])
# access the list key with get method
print(data.get("list"))
```

- Here, with first print statement, we can call direclty with key but put it in [ ] brackets, since dictionary doesn't allow indexing based sequencing
- In second print statement, we used .get( ) method which is an ideal way to fetch the data

### Comparision between two dictionaries

```python
dict1 = {'India': 'New Delhi', 'Japan': 'Tokyo', 'France': 'Paris'}
dict2 = {'France': 'Paris', 'India': 'New Delhi', 'Japan': 'Tokyo'}
dict3 = {'game': 'fottball', 'color': 'black', 'car': 'Land Rover'}
```

- Dictionary supports only comarison operators (==, !=), no relationship operator supported

```python
# check if dict1 and dict2 are identical
dict1 == dict2

# Here, key:value are different in dict1 & dict2, but still Python can recognize this and consider these as identical and display True
# Dictionary comparisons check content equality, not insertion order
```

```python
dict2 == dict3 # not identical, hece False
```

```python
dict1 != dict3 # these two dict are not identical so dict1 is not equal to dict3, hence True
```

# get and setdefault method

-  Dictionary has another method called `setdefault` which also gives the value of key but if no keys are there then it will add that key with value
- But if key is already there, then it will return the default value from dict, and given value will be ignored
- syntax
  - dict_name.setdefault(key, default)

```python
new_dict = {'name': 'mayank', 'age': 12, 'class': 7}
new_dict.get('name', 'sam') # check the value for name key, if key is there then return the value of that key from dict, sam will be ignored
```

```python
new_dict.get('school', 'mdn') # check the value for name key, if key is there then return the value of that key from dict, else mdn will be displayed
```

```python
# stil we have same dict, no changes
new_dict
```

- But if we use setdefault method, then it will set the key if not present, then it will add key:value, and if the key is there, then it will just return the value from the dict, and the passed values will be ignored

```python
new_dict.setdefault('name', 'sam')
```

```python
new_dict.setdefault('school', 'mdn')
```

# Dictionary UNION Operator

- Union Operator types:
  - Vertical pipe operator, denoted as `|`, which merges two dictionaries as a result, creating a new dictionary, leaving the original unchanged.
    - (If both dictionaries have the same key, the right operand's value takes precedence in the merged result)
  - Vertical equal pipe operator, denoted as `|=`, merges two dictionaries but modifies the left dictionary in place; the right dictionary will remain unchanged.

```python
math_ops1 = {'mul': 'x', 'div': '/'}
math_ops1
```

```python
math_ops2 = {'add': '+', 'sub': '-'}
math_ops2
```

```python
# merge both dictionnary with vertical union operator
final_ops = math_ops1 | math_ops2
final_ops
```

```python
print(math_ops1)
print(math_ops2)
print(final_ops)
```

- Here, we can see | operator merged the two dictonary and no changes in orginal dictionary rather created a new dictionary
- Now, try with two new dictionary where we have common key:value pair

```python
roman_num1 = {'I': 1, 'II': 2, 'III': 3}
roman_num1
```

```python
roman_num2 = {'IV': 4, 'V': 5}
roman_num2
```

```python
# merge with |= operator which will modify the roman_num1 in place, but roman_num2 will be same as it was
roman_num1 |= roman_num2 # we can't assign this to a new string or variable, it will give a syntax error

print(roman_num1)
print(roman_num2)
```

# Dictionary Comprehensions

- A Quick way to create a dictionaries using single line (functional prgramming)
- It supports filtering and mapping, commonly used for map a dictionary to another
- Syntax:
  - {key: value for item in iterable}

```python
nums = [1, 2, 3, 4]

squares = {x: x**2 for x in nums}

print(squares)
```

```python
# with condition filter
nums = [1, 2, 3, 4, 5]

even_squares = {x: x**2 for x in nums if x % 2 == 0}

print(even_squares)
```

```python
# modify existing dictionary
d = {'a': 1, 'b': 2, 'c': 3}

updated = {k: v * 10 for k, v in d.items()}

print(updated)
```

```python
# Swap keys and values
d = {'a': 1, 'b': 2}

swapped = {v: k for k, v in d.items()}

print(swapped)
```

```python
# Create Dict from Two Lists with zip function
keys = ['a', 'b', 'c']
values = [1, 2, 3]

d = {k: v for k, v in zip(keys, values)}

print(d)
```
