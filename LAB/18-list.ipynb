#### Why did the list_1 < list_2 not give an error here, since we have a string in list_2, but when we try to compare the second list, it gave me a TypeError.
- Here are the reasons:
  - It compares the elements to elements from each list
  - In the first case, 10 from list_1 is not greater than 40 from list_2, so it found the result and never compares the rest of the elements
  - In the second case, it compares a to 1, which is not possible, so it gives an error

  - 👉 Python is lazy in comparison. It stops as soon as it can decide

# List Comprehension

- List comprehension is a short and clean way to create lists in Python.
- Instead of writing a full for loop, you can write everything in one line
- Basic Syntax
  - [expression for item in iterable]

```python
# Without list comprehension
squares = []
for i in range(5):
    squares.append(i**2)

print(squares)
```

```python
# With list comprehension (for loop)
squares = [i**2 for i in range(5)]
print(squares)
```

```python
# with list comprehensions (with if condition)
even_numbers = [i for i in range(10) if i % 2 == 0]
print(even_numbers)
```

```python
# with list comprehensions (with if-else condition)
result = ["Even" if i % 2 == 0 else "Odd" for i in range(5)]
print(result)
```

```python
# with list comprehensions (with strings)
chars = [ch.upper() for ch in "python"]
print(chars)
```

```python
# with list comprehenstions (with nested loop)
pairs = [(x, y) for x in [1, 2] for y in [3, 4]]
print(pairs)
```

### Real use case for List Comprehentions
- Filtering the data
- Data Transformation
- Cleaning data

```python
# filtering
names = ["Amit", "John", "Ankit", "Sara"]
a_names = [name for name in names if name.startswith("A")]
print(a_names)
```

```python
# transforming/transferring (in pipeline, DevOps)
files = ["app.log", "data.txt", "error.log"]
log_files = [f for f in files if f.endswith(".log")]
log_files
```

```python
# cleaning (converting the string into numbers)
numbers = ["1", "2", "3"]
int_numbers = [int(n) for n in numbers]
int_numbers
```

```python
# merging
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print(flat)
```

### Apart from the above operations, List Comprehensions support two functional programming operations:
- Mapping: Transforming elements to a new value
- Filtering: Keeping only elements that match the criteria as per condition

##### Mapping Function

```python
# first way with empty list
list1 = []
print(list1)

# append the items with a for loop
for i in range(1,5):
    list1.append(i)

print(list1)
```

```python
# second way to convert it into a list
list(range(1,5))
```

```python
# third way, as list comprehension 
## recommended way
list2 = []
print(list2)

list2 = [i for i in range(1,5)]
print(list2)
```

```python
# let's map it to a new value and observe a original list will not be modified
print(f'original list is : {list2}')

list_new = [i*2 for i in list2]
print(f'new modifeid list is: {list_new}')

print(f'reference point for orig list -  {id(list2)}')
print(f'reference point for modified list - {id(list_new)}')
```
