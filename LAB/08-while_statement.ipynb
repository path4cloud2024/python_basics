# While loop
- A while loop in Python repeatedly executes a block of code as long as a specified condition is True. It's used when you don't know the exact number of iterations needed in advance. The loop continues to run until the condition becomes False

```python
while 10 < 10:
    print('hello')

print('10 is not less than 10, so while never print')
```

```python
# print the numbers 1 to 10
num = 1
while num <= 10:
    print(num)
    num += 1
```

```python
# print the first power of 7 greater than 1000
num1 = 7
while num1 <= 1000:
    num1 = num1 * 7
num1
```

# Time to check
- Write a program which will ask for a user to give a name and keep on asking until user gives a blank/empty name
- Hint: use while loop and introduce a break statement

```python
## SOLUTION
while True:
    name = input('Enter your name: ').strip() # here, strip will remove the space from begining or ending of given string
    if name == '': # if given string is emtpy
        break
    print(f'Hello, {name}, Welcome to Python Class !!')
```

```python
"""Class average program with sentinel-controlled iteration."""

total = 0  # sum of grades
grade_counter = 0  # number of grades entered

# processing phase
grade = int(input('Enter grade, -1 to end: '))  # get one grade

while grade != -1:
# Here we can skip the assignment for grade before the while loop, and using the 'WALRUS OPERATOR', we can assign the value inside the while loop as well
# while (grade := int(input('Enter grade, -1 to end: '))) != -1:
    total += grade
    grade_counter += 1
    grade = int(input('Enter grade, -1 to end: '))

# termination phase
if grade_counter != 0:
    average = total / grade_counter
    print(f'Class average is {average:.2f}') # here,:.2f is the precision for float value and instructing the print only 2 decimal values for average
else:
    print('No grades were entered')
```
