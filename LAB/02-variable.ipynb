```python
# to define the variable, we use assignment opertor
a = 10 

# here python, consider this as a int datatypes and create this object and assign the memory automatically
# there is Garbage Collector, which can reclaim the assigned memory to objects if not in use.
```

```python
# to display the value of defined vaiable, just call that variable
a
```
