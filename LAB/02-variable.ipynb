# Variables in Python
- It is good to assign some values to a name, and then use that name repeatedly, called **Variables**.
- Variable/name on the left, and a value on the right. The = between them tells Python to take the value on the right and create the object in memory.
- There are some inbuilt variables which start with a capital letter (will look into those later in this section)
- If we have multiple keywords or want to separate those words in a variable, then use underscore(_)

```python
# to define the variable, we use the assignment =  operator
a = 10 

# here python, consider this as an int datatype and create this object, and assign the memory automatically
# There is a Garbage Collector, which can reclaim the assigned memory to objects if not in use.
```

```python
# to display the value of a defined variable, just call that variable
a
```

```python
b = 5
```

### we can do mathematical operation on these variables

```python
a + b
```

```python
# we can store the result in another variable as well
total = a + b
```

```python
total
```
