## Tuple packing & unpacking

- Any sequence of elements can be unpacked by assigning to a comma-separated list of variables
- These are iterable items, so can be unpacked (strings, lists, tuples, range
- Number of variables must match the number of elements, else `ValueError` occurs

#### Packing and Unpacking Tuple

```python
# packing
z = 1,2,3
z
```

```python
# unpacking
a, b, c = z
print(a)
print(b)
print(c)
```

#### Packing and Unpacking Strings

```python
# packing
s1 = 'sam'
s1
```

```python
# unpacking
a, b, c = s1
print(a)
print(b)
print(c)
```

#### Packing and Unpacking list

```python
# packing
l1 = [1, 2, 3]
l1
```

```python
# unpacking
a , b, c = l1
print(a)
print(b)
print(c)
```

#### Packing and Unpacking range

```python
# packing
x = range(1,4)
x
```

```python
# unpacking
a, b, c = x
print(a)
print(b)
print(c)
```

### Pythonic Beauty with tuples

- Swap values of two number without tuples
- Now, use tuple magic to swap those

```python
# Swap without tuples
x = 10
y = 20
print('x is:', x)
print('y is:', y)
temp = x
x = y
y = temp
print('post swap')
print('now x is:', x)
print('now y is:', y)
```

```python
# Swap with tuples
x1 = 30
y1 = 60
print('x1 is:', x1)
print('y1 is:', y1)
x1, y1 = y1, x1
print('post swap')
print('now x1 is:', x1)
print('now y1 is:', y1)
```

## Stared Advanced unpacking with tuple

```python
a, *b = (1, 2, 3, 4) # here, the first element in the tuple will be assigned to a, and the rest are assigned to b as a list

print(a)  # 1
print(b)  # [2, 3, 4]
```

```python
a, *b, c = (1, 2, 3, 4) #here, first element in the tuple will be assigned to a, last element to c, and middle elements to b as a list

print(a)
print(b)
print(c)
```

```python
*a, b, c = (1, 2, 3, 4) #here, the last element from the list goes to b and c, and initials element goes to a as started unpacking
print(a)
print(b)
print(c)
```

- Python allows one starred unpacking in tuple
- Strated variable can be anywhere, while unpacking Python analyze the numnber of variables and then assign the values based on order they defined.

## ENUMERATE function
- It unpacks the items/elements from a list or range and passes those as tuples with it's index value

```python
# Let's say, we have one list of fruits and we want to unpack those
fruits = ['kiwi', 'apple', 'orange', 'grapes']
print(fruits)
print(type(fruits))
```

```python
new_list = list(enumerate(fruits))
print(new_list)
print(type(new_list))
```

```python
# Check the type of first item in this list
type(new_list[0])
```

```python
# Or we can convert it to a tuple
new_tuple = tuple(enumerate(fruits))
print(new_tuple)
print(type(new_tuple))
```

```python
list(enumerate(range(1,5)))
```

```python
# We can do with the range function and enumerate with a for loop
for index, value in enumerate(range(10,60,10)):
    print(f'{index}: {value}')
```

```python
blocks = [
    0x2581, 0x2582, 0x2583, 0x2584,
    0x2585, 0x2586, 0x2587, 0x2588
]

for b in blocks:
    print(f"{hex(b)} → {chr(b)}")
```

# Let's create a Text_based visualization chart using `enumerate` function
- Use Case
  - CLI Dashboard (in Linux)
  - Monitoring scripts
  - Log Analysis
  - Server Incident report
  - Automation report

```python
## With dictionaries
incidents = {
    "Server Down": 15,
    "High CPU": 8,
    "Disk Full": 30,
    "Network Issue": 6
}

print("\nDevOps Incidents:\n")
print(f'Issues{"Count":>15}    Graph')

for issue, count in incidents.items():
    if count < 15:
        print(f'{issue:<15}{count:>5}    {"█" * count}') # while color
    else:
        print(f'{issue:<15}{count:>5}    {"\033[91m█\033[0m" * count}') # red color
```

```python
# With list
number = [10, 15, 7, 21, 3, 28]
print('\nBar Chart')
print(f'Index{"Value":>8}    Bar')
for index, value in enumerate(number):
    print(f'{index:>5}{value:>8}    {chr(0x2588) * value}')
```
