# f-strings
- They are strings. We just define them differently.
- A f-string must have `f` before the opening quote.
- Inside of the f-string, you can have `{}`. They can contain any Python expression like variable name, operation, function call etc.
- The `{}` automatically turn their contents into a string. So you can use integers and other values.

```python
f1 = 10
f2 = 20
print('f1 value is:', f1, 'and f2 value is:', f2) ## without f-strings, normal way

print (f'f1 value is {f1}, and f2 value is {f2}') ## with f-stings, SMART WAY

## can operate as well
print(f'total of {f1} and {f2} is - {f1+f2}')
```

# More on Python Strings

### Create string
`s = "Hello, World!"`  
Creates a simple string variable.

### Multiline strings
`s = """This is a multiline string"""`  
Allows writing text across multiple lines.

### Raw string
`s = r"C:\Users\John"`  
Treats backslashes as normal characters (no escape sequences).

### String repetition
`s = "Hello" * 3`  
Repeats the string → "HelloHelloHello".

### String indexing
`first_char = s[0]`  
Gets the first character of the string.

### String slicing
`substring = s[1:5]`  
Extracts characters from index 1 to 4.

### Reverse string
`reversed_string = s[::-1]`  
Reverses the string.

### Convert to uppercase
`s.upper()`  
Converts all characters to uppercase.

### Convert to lowercase
`s.lower()`  
Converts all characters to lowercase.

### Capitalize string
`s.capitalize()`  
Capitalizes the first letter, rest lowercase.

### Title case
`s.title()`  
Capitalizes the first letter of each word.

### Swap case
`s.swapcase()`  
Converts uppercase to lowercase and vice versa.

### Strip whitespace
`s.strip()`  
Removes spaces from both ends.

### Left strip
`s.lstrip()`  
Removes spaces from the left side.

### Right strip
`s.rstrip()`  
Removes spaces from the right side.

### Replace substring
`s.replace("old", "new")`  
Replaces "old" with "new".

### Split string
`parts = s.split(",")`  
Splits string into a list using a delimiter.

### Join strings
`",".join(["a", "b", "c"])`  
Joins list into a string with a comma.

### Check if string starts with
`s.startswith("Hello")`  
Returns True if string starts with "Hello".

### Check if string ends with
`s.endswith("World!")`  
Returns True if string ends with "World!".

### Find substring
`index = s.find("World")`  
Returns index of substring or -1 if not found.

### Count occurrences
`count = s.count("l")`  
Counts how many times "l" appears.

### Check if string is alphanumeric
`s.isalnum()`  
True if string contains only letters and numbers.

### Check if string is alphabetic
`s.isalpha()`  
True if string contains only letters.

### Check if string is digit
`s.isdigit()`  
True if string contains only numbers.

### Check if string is lowercase
`s.islower()`  
True if all characters are lowercase.

### Check if string is uppercase
`s.isupper()`  
True if all characters are uppercase.

### Check if string is title case
`s.istitle()`  
True if each word starts with uppercase.

### Check if string is whitespace
`s.isspace()`  
True if string contains only spaces or tabs.

### Center string
`s.center(20, "*")`  
Centers the string with padding using "*".
