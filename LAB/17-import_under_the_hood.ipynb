# IMPORT Under the Hood

- We can import modules with keyword `import`
- With . (dot), we can use the defined function or variable within that module
- Example:
  - import module_name
  - module_name.function_name(args)
  - module_name.variable_name
    - import math
    - math.sqrt(14) # function_name
    - math.pi # variable_name
      
- We can import some speicific identifier from a module
  - Example:
  - from module_name import identifier
  - ** this imported identifier can be used without qualifying (with .)
    - from math import sqrt,  pi
    - sqrt(14)
    - pi

- We can also import as alias
  - import math as m
  - m.sqrt(14)
  - m.pi

- We can import all - Wildcard Imports (but not recommended)
- It silently replaces the other identifiers with module identifiers (check below example)
  - from math import *
  - sqrt(14)
  - pi

```python
# lets assign the value to variable e
e = "Hello,, there"
print(e)
```

```python
# now import all
from math import *
print(e)
```

### Check out the datetime module
- Find out the date difference from today towards future data
- Hint - https://docs.python.org/3/library/datetime.html#module-datetime

```python
#Solution
from datetime import datetime

user_input = input("Enter a future date (YYYY-MM-DD): ")
future_date = datetime.strptime(user_input, "%Y-%m-%d")
today = datetime.today()

difference = future_date - today

if difference.days < 0:
    print("Date is in the past!")
else:
    print(f"Days remaining: {difference.days}")
```

```python
# Show locale and time zone
import locale
import datetime

# Get system locale
loc = locale.setlocale(locale.LC_ALL, '')

# Get current date & time with timezone info
now = datetime.datetime.now().astimezone()

# Print results
print("Locale (Language, Encoding):", loc)
print("Time Zone:", now.tzinfo)
print("Current Time:", now)
```
