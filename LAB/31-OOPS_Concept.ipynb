# What is OOPs?

- OOPs (Object-Oriented Programming System) is a programming paradigm where the main focus is on objects rather than just functions and logic.
- Think of objects as real-world things:
  - A Car is an object → it has properties (color, speed, model) and behaviors (drive, brake).
  - A Student is an object → it has properties (name, roll number, age) and behaviors (study, write exam).
- OOPs allows us to model such real-world entities in our programs, making them easy to understand, reuse, and maintain.

### Core Concepts of OOPS

1. Class
- A class is like a blueprint or template. It does not occupy memory directly. It just defines what properties and behaviors objects created from it will have.
- Example:
  - A Car class defines what every car should have (engine, color, brand) and what it can do (drive, stop).

2. Object
- An object is an instance of a class. When you create an object, you bring the class to life.
- Example:
  - Car myCar = new Car("Red", "Tesla");
  - Here, myCar is an object of the Car class.

3. Encapsulation
- Encapsulation means wrapping data and methods together inside a class and controlling access to them. This ensures data security and avoids accidental modifications.
- Example:
  - A bank account object hides its balance. You cannot change it directly; you can only deposit or withdraw using the available the methods.

4. Abstraction
- Abstraction means showing only the essential features and hiding unnecessary details.
- Example:
  - When you drive a car, you just use the steering wheel, brakes, and accelerator. You don’t need to know how the engine works internally.

5. Inheritance
- Inheritance allows a class to acquire properties and behaviors of another class. This promotes code reuse.
- Example:
  - A Dog class can inherit from an Animal class.
  - Animal may have properties like name, age and behaviors like eat(), sleep().
  - Dog automatically gets these features and can add more like bark().

6. Polymorphism
- Polymorphism means one thing, many forms. In OOPS, it allows the same function to behave differently depending on the object.
- Example:
  - A method makeSound() could make a Dog bark and a Cat meow.
  - Same function name, different behaviors.

## Why OOPS is Important?
- Reusability → Code can be reused through inheritance.
- Scalability → Easy to extend and modify without breaking everything.
- Security → Encapsulation protects data.
- Real-world modeling → Closer to how humans think about problems.
- Maintenance → Easier to debug and update.

# Classes and Objects in Python

## Class
- In Python, a Class is like a blueprint or a template for creating Objects. Since in Python, everything is considered an object.
- If a "Car" is a class, then your specific car, a "Red Tesla," is an object (an instance) of that class.
- It does not occupy memory directly. It just defines what properties and behaviors objects created from it will have.
## Object
- An object is an instance of a class. When you create an object, you bring the class to life.
- It is a specific instance created from the class blueprint/template.
- Each object has its own unique set of data, and data can be physical data or logical data

```python
class emp:
    ops = "Calculating Salary"
    print(ops)
    
    def data(self,n,r):
        self.name=n
        self.name=r
        print("name:",n,"- role:",r)

    def sal(self,t,a,p):
        self.ctc=t
        self.allow=a
        self.pf=p
        print("salary:",t+a-p)

e1=emp()
e1.data("amit","devops")
e1.sal(23000,7500,5000)
print("*********************")
e2 = emp()
e2.data('sam','HR')
e2.sal(45000,12500,3500)

print('\n')
print("---Class/Object/Methods infomation---")
print('---------------------------------------------------------')
print(f'emp is a: {emp}')
print(f'method resolution order "mro" is a: {emp.mro} and In Python 3, every class you create automatically inherits from object, which provides built-in attributes.')
print(emp.mro()) # to print the MRO order
print(f'emp class has two functions: {emp.data} and {emp.sal}')
print('---------------------------------------------------------')
print(f'ops is a class attributes and can be referenced via class instsance as emp.ops and value is: {emp.ops}')
print(f'ops is a class attribute, and can be referenced via object instance as e1.ops and value is: {e1.ops}')
print('---------------------------------------------------------')
print(f'e1 is object and stored at memory location: {e1}')
print(f'e2 is object and stored at memory location: {e2}')
print('---------------------------------------------------------')
```

- class is a keyword that is used to create a class; emp is a class in the above example.
- Objects are instances created from a class, where all methods defined under the class can be referenced. E1 is an object in the above example
- `__int__` is a special method that we do not call explicitly, and it is used to call the variable
- self is a variable which reference to itself in the current class; it is always the first parameter in a method definition
- Class Attributes: are shared among all objects of the class, must be defined outside of a method, directly in a class
- Instance Attributes: are specific to each object and defined within the __init__ method
