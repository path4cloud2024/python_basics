## Default Parameter Values for modules
- A function can have default parameter values
- While calling a function, if an argument value is omitted, then dafault value is automatically used
- specify the value by following parameter name with = (without space), like lenght=2
- check the example below, where we are defining the new function to calculate the area of a rectangle

```python
def area(length=2, width=4):
    """ To calculate the area of a rectangle. """
    return length * width

area() # without providing any argument value, will use default values from functions
```

```python
# but we can give the values while calling
area(length=3) # with first argument
area(width=5) # with second argument 
area(width=3, length=3) # with both arguments, order doesn't matter here
```

```python
area(3) # here by default, its value is assigned to the first keyword argument as positional arguments
```

## Arbitrary Arguments Lists
- A function that can receive any number of positional arguments
- Example, built-in functions `min` and `max` have at least two required parameters and `*args` for more
- Here, * tells to unpack the given values and assign those to function one-by-one
- Check out the example below

```python
def average(*args):
    """ calculating the average """
    return sum(args) / len(args)

average(1, 4, 7, 9, 4)
```

```python
# Python can accept the iterable value and unpack those at the time of function call
marks = [98, 91, 89, 88] # it is a list
average(*marks) # *args is an identifier, but any identifier can be used, since in this case we used *marks
```

# Time to check
- Create a function named product that receives at least one argument and returns the product of all its arguments.
- Call the function with 10, 20, and 30
- Then call with the sequence of integers produced by range(1, 10, 2)

```python
#Solution
def product(num,*args):
    """ Calculating the product """
    product = num
    for values in args:
        product *= values
    return product

## calling function
product(2) # with one argument
```

```python
# with given list
product(10, 20, 30)
```

```python
# with range
product(*range(1, 10, 2))
```

# Identifier SCOPE RULES

- Each identifier has a scope that determines where you can use it in a program.
- Normally, this is the heirarchy, we follows:
  - Local: look into local body of function
  - Enclosing: look into nested fucntion of scope
  - Global: Outside of any function/class (can be used anywhere in .py file)
  - Built-ins

```python
# Lets define a global variable
glb = 10
# from now onwards, this glb has a value of 10 which is globally assigned
```

```python
# lets call global variable
glb
```

```python
# lets try to call this inside the fucntion
def try_global():
    print('current value of glb is: ', glb)
try_global()
```

```python
# Now, let's define the same variable but inside the function body, which does not affect the global value. We can prove
def modify_global():
    glb = 5
    print('current value of glb is:', glb)
modify_global()
```

```python
# lets prove the global value of glb
glb
```

```python
# but it's possible to modify the global variable from inside a function, we need to explicitly reference to gloabl variable, and it will modify it
def change_global():
    global glb
    glb = 15
    print('current value of glb is:', glb)
change_global()
```

```python
# Lets prove that it actually modified the global values of variable
glb
```

## LIST
- Ordered (keeps the order of insertion).
- Mutable (you can change elements).
- Allows duplicates.
- denotes with [ ]

```python
fruits = ["apple", "banana", "apple", "mango"]
print(fruits)          # ['apple', 'banana', 'apple', 'mango']
fruits.append("orange") # add
fruits[1] = "grapes"    # modify
print(fruits)          # ['apple', 'grapes', 'apple', 'mango', 'orange']
```

## TUPLE

```python
coordinates = (10, 20, 10)
print(coordinates)      # (10, 20, 10)
# coordinates[0] = 5    # ❌ Error: tuples are immutable
```

## SET

```python
unique_numbers = {1, 2, 3, 3, 4}
print(unique_numbers)   # {1, 2, 3, 4}  (duplicate 3 removed)

unique_numbers.add(5)
print(unique_numbers)   # {1, 2, 3, 4, 5}
```
