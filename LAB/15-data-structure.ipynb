## LIST

```python
fruits = ["apple", "banana", "apple", "mango"]
print(fruits)          # ['apple', 'banana', 'apple', 'mango']
fruits.append("orange") # add
fruits[1] = "grapes"    # modify
print(fruits)          # ['apple', 'grapes', 'apple', 'mango', 'orange']
```

## TUPLE

```python
coordinates = (10, 20, 10)
print(coordinates)      # (10, 20, 10)
# coordinates[0] = 5    # ❌ Error: tuples are immutable
```

## SET

```python
unique_numbers = {1, 2, 3, 3, 4}
print(unique_numbers)   # {1, 2, 3, 4}  (duplicate 3 removed)

unique_numbers.add(5)
print(unique_numbers)   # {1, 2, 3, 4, 5}
```

## DIctionary

```python
person = {"name": "Amit", "age": 30, "city": "Pune"}
print(person["name"])   # Amit

person["age"] = 31      # update
person["job"] = "DevOps" # add new
print(person)
# {'name': 'Amit', 'age': 31, 'city': 'Pune', 'job': 'DevOps'}
```
