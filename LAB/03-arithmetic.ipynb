### expression will be evaluated always from righ to left if no parentheses else, place of parentheses will evaluate first

```python
4 ** 3 ** 2 # order will right to left

# 4 ** 9
```

```python
(4 ** 3) ** 2 # here, first parentheses from left side
# 64 * 2
```

```python
4 ** (3 **2) # here, first parentheses from right side
# 4 ** 9
```

##### when we have multiple operators of the same level of precedence in expression then it will get evaluated from left to right
