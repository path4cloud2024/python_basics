### Input function stores the value as string
- here is the proof, below example
- it is called string concatenation

```python
value1 = input("enter first number = ")
```

```python
value2 = input("enter second number = ")
```

```python
value1 + value2 # we expect 15 but it won't since number 10 and number 5 are stored as string, you can print value1 and check
```

```python
value1 # here output will be enclosed with single quotes since it is stored as string
```

### to get the input as string, we can convert into a right format

```python
value3 = input("enter any value: ")
```

```python
value3 # still print it as string
```

```python
value3 = int(value3) # it will conver the string into int datatype and store the new value for value3 variable
```

```python
value3
```

# Comparison Operator
- Before we have seen + operator takes two values and return the new one. If two interger then new interger, if two strings then new strings
- And, = operators assign the value of its right to the left variable
- But we have other operators which are completly different, like == which compares the two variables/things whether those are equal or not and returen **TRUE** or **FALSE**

```python
## example to check the values with comparison == operator
100 == 100
```

```python
abc = abc #will give error since wihtout quotes, it will be considered as defined variable
```

```python
'abc' == 'abc'
```

```python
'abc' == 'def'
```

# Other Comparison Operator
- '!=' -- this means "not equal" (opposite of ==)
- '<' -- less than
- '>' -- greate than
- '<=' -- less than or equal
- '>=' -- greater than or equal

```python
# != no equal example
x = 10
y = 20
x != y
```

```python
# < less than 
x < y
```

```python
# > greater than example
x > y
```

```python
# We can compare the strings as well
'abc' < 'def' # it will be true since abc comes before def
```
