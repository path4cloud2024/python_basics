# How can we get an integer from a string?
- `input` provides us with a vluw which always be a string. But what if we want to treat it as an integer?
- We can invoke `int()` on a string, to get a new integer back based on given value.

```python
a = '123'
type(a) # here string is given and check the type - it says its type is STR
```

```python
a1 = int(a)
type(a1) # here, we are converting the value of a into an int and checking its type
```

```python
# we can convert the number while getting input from user
a2 = input('enter the number: ')
n = int(a2)
print(f'you entered the number {n}')
```

# Time to check
- Define one secret number between 1 to 100
- Ask user to give number in this range
- Check the input number and print if its correct guess or low or high 

```python
## SOLUTION
secret = 25
guess = int(input('Enter the number from 1 to 100: ')) # If we don't invoke the int here, then it will give an error since the default is string.
if guess == secret:
    print('You got it !!')
elif guess < secret:
    print('You are below, try again')
else:
    print('You are above, try again')
```

# Time to check
- Get the string from user
- Get the index value from user
- Print the character at given index if index is out of range then print error

```python
## SOLUTION
string = input('enter any word: ')
index = int(input('enter an index value: '))
if index >= len(string):
    print('Index value is out of range')
elif index < 0:
    print('You entered the negative value')
else:
    value = string[index]
    print(f'Value of index {index} in {string} is {value}')
```

## Searcching in Strings
- To search for a smaller string in a given string, use the `in` operator which is going to return `TRUE` or `FALSE`
- Search from left side to right side

```python
a in abcd # here is the proof, string must be quoted
```

```python
'a' in 'abcd'
```

```python
'e' in 'abcd'
```

```python
'bd' in 'abcd' # here it will not look for both character at different place rather it consider as one word like bd only
```

```python
'bd' in 'abcbd'
```

## Time to check
- Write a code to check the given string, start with vowel(a,e,i,o,u) and if yes then add V in front of it
- Else, add C at the end of given string

```python
## SOLUTION
word = input('enter the any word: ')
if word[0] in 'aeiou':
    print(f'V{word}')
else:
    print(f'{word}C')
```

## Using methods on strings

```python
# checking the given string whether its number, alphabetic or alphanumeric
str1 = input('Enter your input ')
if str1.isdigit():
    print('It is numeric')
elif str1.isalpha():
    print('It is alphabetic')
elif str1.isalnum():
    print('It is alphanumeric')
else:
    print('It contains special symbol')
```
