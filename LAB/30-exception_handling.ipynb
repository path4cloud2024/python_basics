# Exception Handling

- Exception handling is a mechanism that allows your program to respond to unexpected events (like invalid input or missing files) without crashing. Instead of terminating abruptly, Python lets you detect the problem, respond to it, and continue execution when possible.
- In Python, this is managed through a structured flow using four main keywords.

#### Syntax of Exception Handling

- Python's error management follows a specific order of execution for handling exceptions: try, except, else and finally

- try: Wraps the code that might trigger an error.
- except: Catches and handles specific errors if they occur.
- else: Executes only if the try block succeeds without any exceptions.
- finally: Runs no matter what, making it perfect for "cleanup" tasks like closing files or database connections. 

```python
try:
    result = 100 / 0
    
except ZeroDivisionError:
    print("You can't divide by zero!")
 
else:
    print("Result is", result)
    
finally:
    print("Execution complete.")
```

#### Different Types of Error:

- Syntax error: issue in logic or indentation or invalid index, occurs at complie time.
- Runtime error: occur at runtime and can be managed using exception handling (e.g., invalid input, missing files).

### Commonly Encountered Exceptions

- ZeroDivisionError: occurs when you attempt to divide a number by zero.
- FloatingPointError: occurs when a floating-point calculation fails. By default, Python handles most floating-point issues silently.
- EOFError: occurs when input() function reaches end of file condition without reading my data.
- IOError: occurs when an input or output operation failed, it raises when the file opened is not found or when wrtiing data disk is full. **Alias for OSError**
- ImportError: Raised when there is a problem with an import statement.
- IndexError: occurs when a sequence index is out of range.
- KeyError: when a mapping key (dictionary) is not found in the set of existing key.
- NameError: when an identifier is not found locally or globally.
- RuntimeError: Raised when an error occurs that doesn’t fall under other categories.
- ValueError: Raised when a function receives an argument of the correct type but with an invalid value (e.g., converting "abc" to an integer).
- TypeError: Raised when an operation or function is applied to an object of inappropriate type.
- ArithmeticError: class is the base for all errors related to mathematical operations. You don’t usually raise it directly, but it provides a way to catch all math-related errors in one block.
- OverflowError: occurs when the result of a numerical operation is too large for Python to represent. While it handles large integers well, certain floating-point operations (like very large exponentials) can still cause this error.
- AssertionError: raised when the assert statement fails. The assert keyword is often used for debugging or testing assumptions in code.
- AttributeError: occurs when you try to access or assign an attribute that does not exist for an object.
- MemoryError: occurs when Python cannot allocate enough memory for an operation. This usually happens when trying to create extremely large data structures.
- ModuleNotFoundError: raised when a module cannot be found.
- StopIteration: raised when the next() function is called but the iterator has no more items.
- KeyboardInterrupt: raised when the user interrupts the program’s execution (e.g., pressing Ctrl+C).
- SystemExit: raised when sys.exit() is called to terminate the program.
- NotImplementedError: raised when a method that should be implemented in a subclass is not implemented.
- IndentationError: raised when indentation is not correct in Python code.
- TabError:	raised when there is inconsistent use of tabs and spaces in indentation.
- UnicodeError:	raised when encoding or decoding Unicode text fails.

#### 2. Catching Multiple Exceptions
- We can catch multiple exceptions in a single block if we need to handle them in the same way or we can separate them if different types of exceptions require different handling.

```python
a = ["10", "twenty", 30]
try:
    total = int(a[0]) + int(a[1])  
except (ValueError, TypeError) as e:
    print("Data Error:", e)
except IndexError:
    print("List is too short!, index is out of range")
else:
    print(f"Success! The total is: {total}")
```

#### 3. Catch-All Handlers and Their Risks
- Catch-all handler is used to call to catch any exception (similar to else statement). Use only except keyword to define it.

```python
try:
    total = '100' + 100
    print(total)

except:
    print("Something went wrong!")
```

#### 4. Custom Exception
- Catch any exception with `Exception` keyword and print the message

```python
try:
    total = '100' + 100
    print(total)

except Exception:
    print("Something went wrong!")
```

#### 5. Raising an custom exception
- We can define our own custom message if required, can be defined with raise keyword followed by an instance of the exception class
- Basic Syntax:
  - raise ExceptionType("Error message")

```python
# Exception Class
def register_user(username, password):
    if len(password) < 8:
        # You choose when to crash the code based on your own rules
        raise ValueError("Security Alert: Password must be at least 8 characters!")
    print(f"User {username} created successfully.")

try:
    register_user("tech_pro", "123")
except ValueError as error:
    print(f"Registration Failed: {error}")
```

```python
def delete_database(user_role):
    if user_role != "admin":
        raise PermissionError("Access Denied: Only admins can delete data.")
    print("Database deleted.")

try:
    delete_database("guest")
except PermissionError as e:
    print(f"Error: {e}")
```
