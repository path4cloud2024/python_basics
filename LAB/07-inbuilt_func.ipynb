## minimum and Maximum Inbuilt Functions

```python
num1 = int(input("enter first number: "))
```

```python
num2 = int(input("enter second number: "))
```

```python
num3 = int(input("enter third number: "))
```

```python
minimum = num1
if num2 < minimum:
    minimum = num2

if num3 < minimum:
    minimum = num3

print('minimum number from given numbers is: ', minimum)
```
