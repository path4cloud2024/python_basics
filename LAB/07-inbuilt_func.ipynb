# Python Built-in Functions

Python provides many built-in functions that are always available and help us perform common tasks easily.

---

## Numeric Functions

### abs()
`abs(-10)`  
Returns the absolute (positive) value → 10

### round()
`round(3.6)`  
Rounds the number → 4

### pow()
`pow(2, 3)`  
Returns 2^3 → 8

### min()
`min(3, 1, 5)`  
Returns the smallest value → 1

### max()
`max(3, 1, 5)`  
Returns the largest value → 5

### sum()
`sum([1, 2, 3])`  
Returns total → 6

---

## Type Conversion Functions

### int()
`int("10")`  
Converts string to integer → 10

### float()
`float("10.5")`  
Converts string to float → 10.5

### str()
`str(100)`  
Converts number to string → "100"

### list()
`list("abc")`  
Converts string to list → ['a', 'b', 'c']

### tuple()
`tuple([1, 2, 3])`  
Converts list to tuple → (1, 2, 3)

### set()
`set([1, 2, 2, 3])`  
Removes duplicates → {1, 2, 3}

---

## Iteration Functions

### len()
`len("hello")`  
Returns length → 5

### range()
`range(5)`  
Generates numbers from 0 to 4

### enumerate()
`list(enumerate(['a', 'b']))`  
Returns index + value → [(0, 'a'), (1, 'b')]

### zip()
`list(zip([1,2], ['a','b']))`  
Combines lists → [(1, 'a'), (2, 'b')]

---

## Logical Functions

### all()
`all([True, True, False])`  
Returns False if any value is False

### any()
`any([False, False, True])`  
Returns True if any value is True

---

## Input / Output Functions

### input()
`input("Enter name: ")`  
Takes user input as string

### print()
`print("Hello")`  
Displays output on screen

---

## Type Checking

### type()
`type(10)`  
Returns data type → int

### isinstance()
`isinstance(10, int)`  
Checks type → True

---

## Advanced Useful Functions

### sorted()
`sorted([3,1,2])`  
Returns sorted list → [1,2,3]

### reversed()
`list(reversed([1,2,3]))`  
Reverses sequence → [3,2,1]

### map()
`list(map(int, ["1","2"]))`  
Applies function → [1,2]

### filter()
`list(filter(lambda x: x>2, [1,2,3,4]))`  
Filters values → [3,4]

## minimum and Maximum Inbuilt Functions

## write a code first to find out the minimum or maximum value without `inbuilt functions`

```python
num1 = int(input("enter first number: "))
num2 = int(input("enter second number: "))
num3 = int(input("enter third number: "))

## checking minimum number without inbuilt functions
minimum = num1
if num2 < minimum:
    minimum = num2

if num3 < minimum:
    minimum = num3

print('minimum number from given numbers is: ', minimum)
```
