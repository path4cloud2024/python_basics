# Handling Excel file with xlsxwriter module

- XlsxWriter is a high-performance Python module used exclusively for creating and writing Excel files in the .xlsx format. It is widely favored for generating reports because it supports advanced Excel features - like charts, conditional formatting, and formulas - while being much faster and more memory-efficient than alternatives like openpyxl
- Follow for more info - https://xlsxwriter.readthedocs.io/

```python
# import the module
import xlsxwriter
```

```python
# 1. Create a workbook and add a worksheet
workbook = xlsxwriter.Workbook('my_report.xlsx')
worksheet = workbook.add_worksheet()

# 2. Write data (can use cell notation 'A1' or row/col index 0,0)
worksheet.write('A1', 'Item Name')
worksheet.write(0, 1, 'Quantity')   # will write in B1 cell
worksheet.write(0, 2, 'Cost')   # will write in C1 cell

# 3. Add formatting
bold = workbook.add_format({'bold': True, 'font_color': 'blue'})
worksheet.write('A2', 'Apples', bold)
worksheet.write('B2', 50)
worksheet.write(1, 2, 100)   # will write in C2 cell

# 4. ALWAYS close the workbook to save it
workbook.close()
```

# How to Modify an Existing Excel File

- XlsxWriter cannot modify existing files—it can only create new ones from scratch. If you use it on an existing file, it will overwrite everything.
To update an existing sheet, change its name, and add rows, you should use the openpyxl library.
- Follow for more info:
  - https://openpyxl.readthedocs.io/en/stable/
  - https://www.geeksforgeeks.org/python/introduction-to-python-openpyxl/

```python
# import the module
import openpyxl
```

```python
from openpyxl import load_workbook

# 1. Load the existing file
file_path = 'my_report.xlsx'
wb = load_workbook(file_path)  # Opens your existing .xlsx file without deleting its contents.

# 2. Rename the worksheet (assumes modifying the active/first sheet)
ws = wb.active
ws.title = "fruits"  # Changes sheet name to 'fruits'

# 3. Add new rows of data
new_rows = [
    ['Apple', 50, 'Red'],
    ['Banana', 30, 'Yellow'],
    ['Mango', 20, 'Orange']
]

for row in new_rows:
    ws.append(row)  # Automatically adds data to the next available empty row

# 4. Save the changes back to the same file
wb.save(file_path) # Commits your changes to the disk. Always close the Excel file on your computer before running this, or you will get a PermissionError.
print("Sheet renamed and rows added successfully!")
```
