# TIP: If iteration items to a function supports mutable then it can modify the obejct reference in memory
- Like for loop, can iterate over the list and modify in memory directly since list supports mutable feature
- Tuples are immutable so we can't modify them

```python
def cube_func(items):
    'Calculates the cubes of given numbers or a list.'
    for i in range(len(items)):
        items[i] **= 3 
```

```python
cube_list = [1, 2, 3, 4, 5]
id(cube_list)
```

```python
id(cube_list[1]) # make a note of reference point of element at index 1 before we call a function
```

```python
cube_func(cube_list)
cube_list # functions have modified the elements for this list
```

```python
id(cube_list) #but list object still refernce the same point in memory
```

```python
id(cube_list[1]) # but now the reference point for index 1 is changed since the value is changed
```

```python
# Try with Tuples
cube_tuple = (1, 2, 3, 4, 5)
cube_func(cube_tuple)
cube_tuple
```

## implications of passing mutable lists to functions
- When you pass a list to a function, you're passing a reference to the same object, not a new copy, the original list is modified.
- In-place changes affect the original list. Operations like: append(), extend(), remove(), sort()
- Functions that modify input lists can create unexpected behavior.

## Avoiding unintended mutations
- Pass a copy
- Use slicing
- Return a new list (functional style)

#### USE CASE:
- If we are not sure whether a particular number is available in list or not, first we check and if it does, then we can find the index value of that element. This way, we can ignore the unwanted error (again- Do it SMART WAY, Not HARD WAY 😀)

```python
key = int(input('enter value to check:'))
index_value = num_1.index(key) if key in num_1 else None
if index_value is not None:
    print(f'Value {key} is at index {index_value}')
else:
    print(f'Value {key} is not in list')
```

## Python supports the STACK operations with a list
- Stack is also a constrained list where push adds to the top, and pop removes from the top. Basically, follow LIFO (last-in, first-out)
- So, in Python, lists also supports pop() method, which doesn't take any argument and can be used to remove and return the last item in the list
- To push, we have the append() method with a list to add the item to the top

```python
stack_list = []
stack_list
```

```python
# Let's push few items with append() menthod
stack_list.append('first')
```

```python
stack_list.append('second')
stack_list # now, we have few items pushed
```

```python
# Let's try pop() method to remove
stack_list.pop() # no need to give any item or argument or index value, it follows LIFO order
```

```python
# now check the list and we are left with only one item
stack_list
```

- We discussed the Match Case pattern matching with if conditions
- Match can support the list and tuples pattern as well

```python
# define a function which taken arbitory arguments and matches the value
# get the rolled dice value and print
def my_match(*values):
    match values:
        case[1, 1]:
            print('rolled out 1 twice')
        case[6, 6]:
            print('rolled out 6 twice')
        case[x, y] if x == y:
            print(f'rolled out {x} twice')
        case[x, y] if x <= 6 and y <= 6:
            print(f'rolled out {x} and {y}')
        case _:
            print('Invalid values!!!! Please pass only interger upto 6')
```

```python
my_match(10, 3)
```
