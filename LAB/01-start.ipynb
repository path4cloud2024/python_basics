# What is Jupyter?
- Traditionally, we run the code .py file and execute it, but it's not good when we want to try a piece of code.
- It is known as a REPL environment -- READ. EVALUATION, PRINT, LOOP.
- It is best for experimentation and description of your code.
- In this notebook, we can have Python code, markdown documents, visuals (image + chart)

# Handy commands for Jupyter
### This rectangle text box is called "cells" where we type our code or texts. There are two modes
- EDIT mode, which is the default mode of the cell. To get into it, press ENTER.
- COMMAND mode, to get into command mode, either right-click on the cell or press ESC, and then we can pass commands below to Jupyter to get the result as per the command
    - c -- copy the current cell
    - x -- cut the current cell
    - v -- paste the copied or cut cell
    - a -- add a new empty cell ABOVE the current selected cell
    - b -- add a new empty cell BELOW the current selected cell
    - y -- make the current cell sensitive to Python Code and able to run it
    - m -- make the current cell for Markdown Documents

To reset the shell result:
- right click and select "Clear Outputs of All Shell" (this option will clear the result, but won't reset the memory)
- or go to kernel menu and click on "Restart Kernel and Clear Outputs of All Shells" (this option will reset the memory as well)

To rerun all the shell in one go
- go to Run menu and click on "Run all Shells"
- go to kernel menu and click on " Restart Kernel and Run All Cells.."

- Here, print is a verb that functions and can print on the screen.
- Any function in Python, we must use (), right after the function name.

- Argument can be a text value, known as **"STRING"**, and it must be quoted with single or double quotes around it. If not quoted then it will be considered as **VARIABLES** in Python.
- It can print the numbers (integer), which don't need to be quoted 

```python
print(1234567890) # integer number need not be quoted
```
