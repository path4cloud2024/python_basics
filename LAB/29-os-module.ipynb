```python
import xlsxwriter

# 1. Create a workbook and add a worksheet
workbook = xlsxwriter.Workbook('my_report.xlsx')
worksheet = workbook.add_worksheet()

# 2. Write data (can use cell notation 'A1' or row/col index 0,0)
worksheet.write('A1', 'Item Name')
worksheet.write(0, 1, 'Quantity')

# 3. Add formatting
bold = workbook.add_format({'bold': True, 'font_color': 'blue'})
worksheet.write('A2', 'Apples', bold)
worksheet.write('B2', 50)

# 4. ALWAYS close the workbook to save it
workbook.close()
```
