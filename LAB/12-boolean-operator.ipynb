# and operator
- where both condition must be true

```python
nationality = "Indian"
age = int(input("enter your age: "))
if nationality == "Indian" and age >= 18:
    print("eligible to cast your vote")
else:
    print("you are not adult yet")
```

# or operator
- where either of condition is true

```python
weight = int(input("enter your weight in kg: "))
height = int(input("enter your height in cm: "))
if weight >= 54 or height >= 145:
    print("you can participate")
else:
    print("you are not eligible")
```

# not operator
- it is a unary operator which means it takes only one operand and that inverts the truth value of a boolean expression. It returns True if the expression is False, and False if the expression is True
- it is used to check if a condition is not met
- not True evaluates to False
- not False evaluates to True

```python
is_raining = input("Type Yes if its raining outside: ")
if not is_raining == "Yes":
    print("I can go for a walk")
else:
    print("its raining so stay home")
```
