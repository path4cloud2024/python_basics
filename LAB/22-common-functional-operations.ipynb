### map( ) function
- It applies a function to each element in a sequence and returns a new sequence containing the transformed elements. This map function has the following syntax:
  - map(function, iterabel)
  - The iterable items can be a list, tuple, or any other iterable objects like string.

```python
# let's create a square function
def square(value):
    return value ** 2

# call the function
print(square(5))
```

#### Here, we are good if we have one item to pass to a function but what if we have list or any other iterabel items.

```python
# try with this list
list_num = [1, 2, 3, 4, 5]
print(square(list_num))
```

#### Obviously, we can see, our defined function can't iterate over the list since, it is expecting only one argument. But we can overcome this problem with map( ) function

```python
map(square, list_num) # will produce the map object but we can convert it to list and get the result OR use next( ) function to print the result one by one
```

```python
print(list(map(square, list_num)))
```

```python
result = map(square, list_num)
print(next(result))
print(next(result))
```

```python
# we can get the result with for loop as well
for i in result:
    print(i)
```

```python
# Let's try with a while loop
while True:
    try:
        value = next(result)
        print(value)
    except StopIteration: # this is the exception once iterable items are over, and internally hit this exception and break by default
        print('Already printed above with for loop, so nothing is left to print')
        break
```

```python
# Let's calculate again and try with a while loop because above we have already exhausted the iteration, and it hit the exception and nothing was printed
result = map(square, list_num)
while True:
    try:
        value = next(result)
        print(value, end=' ')
    except StopIteration: # this is the exception once iterable items are over, and internally hit this exception and break by default
        print('\nI hit the exception so breaking the while loop')
        break
```

```python
print(*map(square, list_num)) # another method to print the map obejct with * which is Iterable Unpacking Operator
```

## Note: map( ) works on numpy arrays and pandas series. Pandas has its own map function: .map( ), .apply( ) and .applymap( )

### filter( ) function

- It can iterate over some sequence of elements or some iterable object, and we look at each value and decide whether or not to keep it. Only the values that match the criteria we specify
- In simple terms, it keeps only elements that satisfy a condition.
  - filter(function, iterable)
- filter() expects: A function that returns: True  → keep element, False → discard element

```python
list_num = [1, 2, 3, 4, 5]
print(filter(lambda x: x % 2 == 0, list_num)) # here we have used a lambda function with an if condition, where we capture only even numbers
print(list(filter(lambda x: x % 2 == 0, list_num)))
```

```python
# Let's try with a function
def is_even(value):
    return value % 2 == 0 #will return TRUE only if the number is even

result = list(filter(is_even, list_num))
print(result)
```

```python
# Let's try with the square function
print(list(filter(square, list_num)))
```

#### Here, instead of calculating the square of list as per defined function, it return the same list. `WHY?`
#### Beacuse we used filter function and it expects boolen output (TRUE/FALSE), so it checked only if a number is squarable or not, if yes then it got TRUE for all numbers and printed the same list

```python
# let's try with another list
list_num1 = [1, 2, 3, 0, -2]
print(list(filter(square, list_num1)))
```

#### Here, 0 is not squarable so it didnot get the TRUE for 0 so skipped that and printed another numbers

## reduce( ) function

- Applies a function cumulatively to reduce an iterable into a single value
- Used to Aggregate data
- Reductions transform a collection of elements into a single value
- Like we have used, len (to get the length), sum (to get the total of elements), min (to get the minimum), max(to get the maximum)
- Custom reductions with functools.reduce

- functools is an in-built modules library which has reduce function to create custom reductions.
- It requires two arguments:
  - A function receiving two values and returning a result
  - An iterable
- The function argument defines how to combine two values
- Functions that receive other function as arguments are called `higher-order functions`

- The reduce function applies the function to the first two elements in the iterable and then applies the function to the result and the next element, and so on.
- Once iterable item is finished, then it returns the final result.

```python
# Let's define a multiple function
def multiply(x, y):
    return x * y

print(multiply(10, 20))
```

#### Let's try with reduce( ) function

```python
from functools import reduce # reduce is part of the functools module, so we need to import this
list_num = [1, 2, 3, 4, 5]
reduce(multiply, list_num)
```

#### Here, why is it printed as 120?
- reduce function takes the first two elements and passes to the multiply function, gets the result, and then the result and the next element passes to the multiply function
  - first iteration (1 * 2 = 2)
  - second iteration (2 * 3 = 6)
  - third iteration (6 * 4 = 24)
  - fourth iteration (24 * 5 = 120)
  - Now, no more elements to pass, so 120 is the final result

#### SMART TIP:
- For common mathematical operations or might be other operations are already defined in Python as module
- In this case, we have operator module which has mul function to get the multiplication

```python
from operator import mul
reduce(mul, list_num)
```

```python
reduce
```

```python
# We can also make use of a lambda function
result1 = reduce(lambda x, y: x * y, list_num)
print(result1)
```

### reduce( ) with third arguments
#### this is a less-known but very important part of reduce()
- Most people ignore the 3rd argument, but it’s super useful.
- 3rd Argument: It is the initial value used before processing the iterable.
- When provided, reduce takes the third argument and first element from the iterable and processes
- If the iterable is empty and the third argument is given, then reduce returns the third argument
- If the iterable is empty and no third argument is given, then a TypeError occurs

```python
# without 3rd argument
result1 = reduce(lambda x, y: x * y, list_num)
print(result1)
```

```python
# with third argument
result1 = reduce(lambda x, y: x * y, list_num, 10)
print(result1)
```

#### Here, why is it printed as 1200?
- reduce function takes the third argument and first element and passes to the multiply function, gets the result, and then the result and the next element passes to the multiply function
  - first iteration (10 * 1 = 10)
  - second iteration (10 * 2 = 20)
  - third iteration (20 * 3 = 60)
  - fourth iteration (60 * 4 = 240)
  - fifth iteration (240 * 5 = 1200)
  - Now, no more elements to pass, so 1200 is the final result

- Handling empty list as third argument

```python
empty_list = []
result1 = reduce(lambda x, y: x * y, list_num, 0)
print(result1)
# first iteration 0 * 1 = 0
# second iteration 0 * 2 = 0
# and so on, so the final result is also 0
# If the third argument was 1, then 1 would have been multiplied, and we would get 120 
```

- Handling sting as third argument with list of string

```python
course = ['Welcome', 'To', 'Python', 'Class']
c_list = reduce(lambda x, y: x + ' ' + y, course, 'Hello!')
print(c_list)
```

- Simple Analogy
  - Without initializer → start from first two values
  - With initializer → start from your custom value

### Golden Rule (VERY IMPORTANT)
- The initializer type must match the operation
  - Operation	   :    Correct initializer
    - Addition	   : 0
    - Multiplication : 1
    - String concat  : ""
    - List merge	   : [ ]

## zip( ) function

- zip( ) function is used to combine iterables into tuples as per sequence of elements
- Usefull for making separate lists into one single tuples
- Syntax:
  - zip(iterable1, iterable2, ...)

```python
names = ["Amit", "Rahul", "John"]
scores = [90, 85, 88]

result = list(zip(names, scores))
print(result)
```

```python
names = ["Amit", "Rahul", "John"]
scores = [90, 85, 88]

[*(zip(names, scores))] # with * iterable operators
```

```python
# works with more than 2 lists as well
names = ["Amit", "Rahul", "John"]
subject = ["English", "English", "English"]
scores = [90, 85, 88]

result = list(zip(names, subject, scores))
print(result)
```

- Here, we can see that for the subject list, we need to repeat the 'English' as the same count as other elements in the names and scores list because, if we don't put the same count, then zip( ) works up to the length of the shortest input or list (it will zip the element for shortest list)
- But we have another method zip_longest( ) from the itertools library
- Let's try with the shortest list and then see how we can overcome that

```python
names = ["Amit", "Rahul", "John"]
subject = ["English"]
scores = [90, 85, 88]

result = list(zip(names, subject, scores))
print(result)
```

- Here, we have only one element in subject list so zipped only one element from each list and rests are skipped
- Import the itertools and solve this purpose

```python
from itertools import zip_longest
names = ["Amit", "Rahul", "John"]
subject = ["English"]
scores = [90, 85, 88]

result = list(zip_longest(names, subject, scores, fillvalue='Englist'))
print(result)
```

- We can invoke the dictory as well instead of list with zip( ) 

```python
keys = ["name", "age", "role"]
values = ["Amit", 25, "DevOps"]

data = dict(zip(keys, values))
print(data)
```

- It can tranpose the matrix into lists of tuples as well

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

result = list(zip(*matrix))
print(result)
```

- Even more helpful in comparing two lists or mapping two lists

```python
# comparing lists
a = [1, 2, 3]
b = [1, 0, 3]

for x, y in zip(a, b):
    print(x == y)
```

```python
# mapping lists
servers = ["web1", "db1", "cache1"]
ips = ["10.0.0.1", "10.0.0.2", "10.0.0.3"]

mapping = dict(zip(servers, ips))
print(mapping)
```

- We have lists of tuples already, then we can `UNZIP` as well with `zip` function into separate tuples

```python
pairs = [('Amit', 90), ('Rahul', 85)]

names, scores = zip(*pairs)

print(names)
print(scores)
```

## iter( ) function

- returns an iterator object from an iterable.
- syntax
  - iter(iterable)

- Let's create a range function and by default print funciton prints it string representation of range
- Now, try to print it with next ( ) but since range object is not iterable so it will thorugh error
- Now, try with iter( ) function and this time, it will print the range

```python
num = range(1, 5)
print(num) # will print the string representation of the range
```

```python
# try with next( ) method
next(num)
```

```python
# this time, it got the iterator for the iterable range(1,4)
iter_num = iter(num)
print(next(iter_num)) # print 1
print(next(iter_num)) # print 2
print(next(iter_num)) # print 3
print(next(iter_num)) # print 4
print(next(iter_num)) # no more iterable elements to raise exception, will not go further
print("stopped already") # never came here so skipped
```

```python
# we can pass this to a for loop 
for i in iter(num):
    print(i)
```
