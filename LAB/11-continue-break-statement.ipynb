# Break statement
- break alter loop's flow of control
- executing break in loop immediately exits that statement

```python
for number in range(100):
    if number == 10:
        break
    print(number, end=' ')
```

# Continue Statement
- Continue altering the loop's flow of control
- Executing continue in a loop skips the remainder of the loop's suite and moves to the next iteration
- In a while, the condition is then tested to determine whether the loop should continue executing
- In a for loop, processes the next item in sequence (if any)

```python
for number in range(10):
    if number == 5:     ## notice: when 5 is matched, it was skipped and continues to print others
        continue
    print(number, end=' ')
```

### More examples with while loop

```python
i = 1

while True:
    print(i, end=' ')
    if i == 3:
        break
    i += 1
```

```python
i = 0

while i < 10:
    i += 1
    if i == 5:
        continue
    print(i, end=' ')
```

# Simple Rule

🔴 break → “EXIT the loop”

🟡 continue → “SKIP this turn, go next”
