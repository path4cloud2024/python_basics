# Break statement
- break alter loop's flow of control
- executing break in loop immediately exits that statement

```python
for number in range(100):
    if number == 10:
        break
    print(number, end=' ')
```

# Continue Statement
- Continue altering the loop's flow of control
- Executing continue in a loop skips the remainder of the loop's suite and moves to the next iteration
- In a while, the condition is then tested to determine whether the loop should continue executing
- In a for loop, processes the next item in sequence (if any)

```python
for number in range(10):
    if number == 5:     ## notice: when 5 is matched, it was skipped and continues to print others
        continue
    print(number, end=' ')
```

### More examples with while loop

```python
i = 1

while True:
    print(i, end=' ')
    if i == 3:
        break
    i += 1
```

```python
i = 0

while i < 10:
    i += 1
    if i == 5:
        continue
    print(i, end=' ')
```

# Simple Rule

🔴 break → “EXIT the loop”

🟡 continue → “SKIP this turn, go next”

# Time to check
- Write a code to sum only the digit from a given value, ask the user to give input
- Go through each character and if:
  - It's a . (dot), then stop the loop and print the sum
  - It's a non-digit other than . (dot), Then skip that and continue to the next character
  - It's a digit, turn it into a `int` and add to sum, print the total, and go to the next
  - Print final SUM

```python
# Solution
sum = 0
text = input('Enter some value: ').strip() # here strip will remove the white spaces
for each in text:
    if each == '.':
        print(f'Find . (dot), so exiting...')
        break
    if not each.isdigit():
        print(f'skipping non-digit {each}')
        continue
    sum += int(each)
    print(f'adding {each}, and now total sum is {sum}')
print(f'final sum is : {sum}')
```
