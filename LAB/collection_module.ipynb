## Counting Unique Words in List

```python
# defining a list called logs
logs = [
    "INFO User logged in",
    "ERROR Database failed",
    "INFO Request processed",
    "WARNING Disk almost full",
    "ERROR Timeout occurred",
    "INFO User logged out"]

word_count = {}                         # empty dictionary to count the status from logs such as INFO, ERROR, WARNING
for status in logs:                     # for loop iterate over the list and pass each element to status as a string
    word = status.split()[0]            # fetching the first word from each line using string.split method
    if word in word_count:              # checking that word is present in the dictionary or not
        word_count[word] += 1           # If that word is present, then update the dictionary with words and their count in key:value format
    else:
        word_count[word] = 1            # If that word is not present, then insert into dictionary and set the count to 1 in key:value format
word_count
```

# In Python, we have a module called Collections, and that has a `Counter` function to check the frequency of any word/text/string
## Counter always behaves like a dictionary, but with a few extra features.
## Technically, it is a subclass of the standard Python dictionary (dict). This means it stores data in key:value pairs, where the key is the item and the value is the count.
- Let's solve the above example with Counter

```python
# As list
from collections import Counter
logs = [
    "INFO User logged in",
    "ERROR Database failed",
    "INFO Request processed",
    "WARNING Disk almost full",
    "ERROR Timeout occurred",
    "INFO User logged out"]
word_list = []                           # define an empty list to append to each iteration from the logs list
for word in logs:                        # for loop iterate over the list and pass each element as a string 
    counter = word.split()[0]            # fetching the first word from each line using string.split method and then the index 0 value 
    word_list.append(counter)            # each time, the first element will be appended to word_list 
print(word_list)                         # display the current list
Counter(word_list)                       # Let's count the frequency of each word in this list
```

### We can hanlde the different Character Case as well with Counter so that it will count the all words as one
### For Example, we have words 'Cloud', 'CLOUD', 'cloud', then we can convert those to upper or lower to make similar then count the frequency

```python
from collections import Counter

case_w = ['Cloud', 'CLOUD', 'cloud']

similar_w = [w.upper() for w in case_w]  # List Comprehensions, WE CAN USE .casefold() as well which is fine that .lower() or .upper()
#similar_w = [re.sub(r'[^\w]', '', w).casefold() for w in words]
count = Counter(similar_w)

print(count)
```

# defaultdict in collections

## A defaultdict automatically creates a default value for a missing key instead of raising KeyError.
- Common Use Cases
    - Grouping data                 --> defaultdict(int)     # counting
    - Counting                      --> defaultdict(list)    # grouping
    - Building nested structures    --> defaultdict(set)     # unique grouping

```python
# Let's try to update the value with a dict
d1 = {'a': 1}
print(d1['a']) # read the key which exits
```

```python
# read the key which doesn't exist
print(d1['b'])       # shows key error since the key doesn't exist
```

- Now, lets read the key with defaultdict which doesn't exists in dictionary, if key is not there, then instead of `keyerror`, it prints `0`

```python
from collections import defaultdict

d2 = {'a':1}

dd = defaultdict(int, d2)

print(dd['a'])   # display the value of a since it is present in the dictionary
print(dd['b'])   # display the value of b as 0 since it is not present in the dictionary and update the dictionary as well
print(dd)        # notice the class of defaultdict, it is int here
```

```python
type(dd)
```

## defaultdict(int) helps us to count the elements in dictionary and update the key with value 0 if does not exists

## defaultdict(list) helps us to group the elements in dictionary. If we have dictionary of tuples and if we have duplicates keys, then it will groups those with key and values as list

```python
from collections import defaultdict

data = [
    ('a', 1),
    ('b', 2),
    ('a', 3),
    ('b', 4),
    ('c', 5)
]

print(type(data[0]))        # shows the type as a tuple of index 0 in this data list

ddl = defaultdict(list)     # defining the ddl as defaultdict with class list, where we group the elements from the data list

for key, value in data:
    ddl[key].append(value)  # append the key and value in defaultdict with class list

print(ddl)                  # notice the class of defaultdict, it is a list
```

## deafultdict(set) helps us to group the elements in a dictionary, but a `unique group`. If we have a dictionary of tuples and have duplicate keys, then it will group unique keys and values as a list, and remove the duplicate elements from each tuple

```python
from collections import defaultdict

data = [
    ('a', 1),
    ('b', 2),
    ('a', 3),
    ('b', 4),
    ('c', 5)
]

print(type(data[0]))       # shows the type as a tuple of index 0 in this data list

dds = defaultdict(set)     # defining the ddl as defaultdict with class set, where we group the unique elements from the data list

for key, value in data:
    dds[key].add(value)    # add the key and value in a defaultdict with class set, remove duplicate keys

print(dds)                 # notice the class of defaultdict, it is a set
```

## orderdict (An OrderedDict preserves insertion order of keys.)
- Important Note (Python 3.7+): Built-in dict already preserves insertion order

```python
from collections import OrderedDict

do = OrderedDict()
do['b'] = 2
do['a'] = 1
do['c'] = 3
do['d'] = 4
print(type(do))
print(do)
```
