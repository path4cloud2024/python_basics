# min and max with `key` arguements

- Let's take a list of fruits and try to compare them. Python compares the string `Lexicographically.`
- It compares the `ASCII` value (integer) of the first character in the string. If the first character is the same, then compare the second character
- We can check the ordinal values with the `ord( )` function
- We can check the character against an interger with the `chr( )` function

#### min( ) function with key argument

```python
fruits= ['apple', 'Banana', 'Peach', 'lichi', 'Grapes']
min(fruits)
```

- Here, Banana is having the minimum value
- Lets check the ordinal values of each characher

```python
# First character from each string
print(f'a is: {ord('a')}, B is {ord('B')}, P is {ord('P')}, l is {ord('l')}, G is {ord('G')}')
```

- We can see, B is having a 66 value in ASCII table which is indeed a lower in this list, so min( ) function printed the Banana

- Now, with key argument, where we can pass the additional method or function to change the type or representation of each element in given list

```python
# We have a few strings starting with lower and a few strings starting with capital letters
# Let's convert those to lower and then find out the minimum
min(fruits, key=lambda s: s.lower())
```

```python
# First character from each string
print(f'a is: {ord('a')}, b is {ord('b')}, p is {ord('p')}, l is {ord('l')}, g is {ord('g')}')
```

- We can see that a is having a 97 value in the ASCII table, which is indeed a lower in this list, so the min( ) function printed the apple

```python
# Let's convert all to capital letters and then find out the minimum
min(fruits, key=lambda s: s.upper())
```

```python
# First character from each string
print(f'A is: {ord('A')}, B is {ord('B')}, P is {ord('P')}, L is {ord('L')}, G is {ord('G')}')
```

- We can see that A is having a 65 value in the ASCII table, which is indeed a lower in this list, so the min( ) function printed the apple

- But in both operations, case of character was changed only while comparing
- It did not change the original list

```python
fruits
```

#### max( ) function with key argument

```python
# without key arguments
max(fruits) # lichi, since its value is 108 which is higher in this case
```

```python
# with key arguments
max(fruits, key=lambda s: s.lower()) # Peach, its value 112 which is higher
```

```python
max(fruits, key=lambda s: s.upper()) # Peach, its value 80 which is higher
```

# sorted( ) with key argument

- It's sorting the elements in ascending order
- It does not modify the original list
- Use `reverse=False` to get the sorting in descending order
- It is different from the .sort( ) method 

```python
list_num = [12, 5, 98, 2, 0, 26]
sorted(list_num)
```

```python
list_neg_num = [-10, 0, -34, -11, -3]
sorted(list_neg_num)
```

- we can pass the key as an argument which tells Python to compare elements

```python
# Ascending order
sorted(list_neg_num, key=abs) # convert negative numbers into positive and then compare
```

```python
# Descending order
sorted(list_neg_num, key=abs, reverse=True)
```

- Try with dictonay and sort the employee data based on their salary

```python
employees = [
    {"name": "Amit", "salary": 50000},
    {"name": "Rahul", "salary": 70000},
    {"name": "John", "salary": 40000}
]
sorted(employees, key=lambda emp: emp["salary"])
```

#### SMART WAY, NOT HARD WAY
- We do have alternative and clearner method to sort the dictinary like data
- `itemgetter` function is there from `operator` method

```python
from operator import itemgetter
sorted(employees, key=itemgetter("salary"))
```
