# If statement syntax

#### - If is a staic keyword, already reserved in Python.
#### - when we start the if statement then we need to define the condition using operator which we discussed above.
#### - after condtion, if statement must be closed with colon (:), called as suite which tells the python to maintain the indendation for further statement after colon
#### - we normaly uses 4 white spaces as space under if statement to maintain the statement
#### - you can keep any number of spaces for indentation but it must be persistent across all statement
#### - we can close or omit the spaces to close the simple if statement

```python
num1 = 10
num2 = 5
```

```python
num1
```

```python
num2
```

```python
if num1 == num2:
    print(num1, 'is equal to', num2)

if num1 != num2:
    print(num1, 'is not equal to', num2)

if num1 > num2:
    print(num1, 'is greater than', num2)

if num1 >= num2:
    print(num1, 'is greater than or equal to', num2)

if num1 < num2:
    print(num1, 'is less than', num2)

if num1 <= num2:
    print(num1,'is less than or equal to', num2)
```

```python
## Conditional Expresssion
marks = 9
result = ('Pass' if marks >=72 else 'Not Pass')
result
```

```python
## elif example
score = int(input("enter your score: "))

if score >= 90:
    print("Gold Medal")
elif score >=75:
    print("Silver Medal")
elif score >=50:
    print("Bronze Medal")
else:
    print("Not Qualified")
```
