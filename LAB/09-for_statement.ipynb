# for loop
- A for loop in Python is used to iterate over a sequence of items, such as a list, tuple, dictionary, set, or string. It's used when you know the exact number of times you want to repeat a block of code.

```python
# if we print without end then it will print the each characater in string in enw line
for alphabets in 'welcome to path4cloud':
    print(alphabets)
```

```python
for alphabets in "welcome to path4cloud":
    print(alphabets, end=" ")  # here, end is a special variable which takes the values a separator or delimiter
                               # and print all statements in the same line, instead of going to a new line
```

```python
# we can use a sep as well to print something in between of each character as delimiter line hypen (-)
# without sep
print("welcome", "to", "path4cloud")
```

```python
# with sep
print("welcome", "to", "path4cloud", sep="-")
```

💡 Simple Rule
- sep → separates multiple values in one print
- end → controls what comes after each print

```python
## We can try `end` with some delimiter within a for loop, but it prints after each character/string/print
for num in range(10, 0, -2):
    print(num, end="-")
```

```python
## To avoid the last delimiter at the end within the for loop, use the MAP function
nums = list(range(10, 0, -2))
print("-".join(map(str, nums)))

## Too complex!!!!, use the sep
```

#### for loop with list (we will learn list in details later)

```python
total = 0
for num in [12, 34, 5, 23, -45]:  # list is a collection of objects and separated by, in [ ]
    total = total +  num   # we can put this as total += num it is called Augment Assignment
print(total)
```

#### with range() function, which is inbuilt function and print sequence of number

```python
range(10)
```

- It represents the sequence, but doesn’t display it directly.
- To print the sequence, convert it to list

```python
list(range(10))
```

```python
## with for loop
for seq in range(10):
    print(seq, end=" ")
```

### format string

```python
# print a sum of 1 to 50 numbers
sum = 0
for i in range(51):
    sum += i

print(f'total sum of 1 to 50 numbers are {sum}')
```

```python
# we can perform the operation as well in f string
num1 = 10
num2 = 2
print(f'{num1} times {num2} is {num1 * num2}')
```
