# Generator Expressions

- It is similar to List Expressions but uses ( ) instead of [ ]
- It creates a generator (iterator) that yields values one at a time
- No need to define a function or use yield
- Uses `Lazy Evaluation`, which means it produces one value at a time on demand.
  `(List Comprehensions use greedy evaluation which means creating all values immediately when executed`
- Can be iterated only once; to iterate again, create a new generator expression

```python
num1 = [1, 2, 3, 4, 5]
# list comprehensions
list_comp = [i *2 for i in num1]
print(list_comp)
type(list_comp)
```

```python
# Generator Expression
gen_exp = (i * 2 for i in num1)
print(gen_exp) # here gen_exp is an object which creates an iterable expression
print(type(gen_exp))
```

### We can use the next( ) function to generate the next item from this Generator Expression
### Whenever we run the next( ) function, it will print the next item in the Iterable sequence
### We can use a for loop as well to print the entire sequence in one go, but if we have already printed a few with the next( ) function, then it will print only the remaining items in sequence

```python
print(next(gen_exp)) # will print first element
print(next(gen_exp)) # will print second element
```

```python
# now print the remaining items with a for loop
for n in gen_exp:
    print(n, end=' ')
```

```python
# We can call only once, this iterable expression, it's not reusable post execution once
print(list(gen_exp))# already executed above, so now empty list
```

```python
# but if we try with next, then it will hit the exception "STOPITERATION"
next(gen_exp)
```

#### Generator Expression supports the If condition as well (little complex)

```python
# Find out the square only for even numbers
for n in (i * 2 for i in num1 if i % 2 == 0):
    print(n, end=' ')
```

## List Comprehensions executes all items at same time so consumes more memory
## Generator Expression yields values one at a time so consumes less memory
- Let's find out the size of these:

```python
import sys
num = range(10_000_000)

list_size = [i * 2 for i in num]   # huge memory bytes
gen_size  = (i * 2 for i in num)   # memory efficient
print(f'size of list_size: {sys.getsizeof(list_size)}')
print(f'size of gen_size: {sys.getsizeof(gen_size)}')
```

# Geneartor Functions
- We can covert the Generator Expression into functions and it uses `yield` instead of `return` to produce the next item on demand
- After each yield, the function suspends execution until the next item is requested
- Use a Generator Function when a generator expression is longer than 80 characters, which improves the readability, making it easy to maintain, modify, and debug
- Calling a generator function returns a generator object that can be iterated over once

```python
# define gen_funciton with yield
def gen_func(values):
    ''' Calculate the square and yield (pauses) the values after one execution
        will print the next item on a demand '''
    for value in values:
        yield value ** 2

result = gen_func(num1) # it will just generate the Generator Object in memory, but will not print any values
print(result)
```

#### To print the value use next( ) function or a for loop

```python
print(next(result))
print(next(result))
```

```python
# remaing items with a for loop
for remaining_item in result:
    print(remaining_item)
```

```python
# if call again, then it hits the exception since it expires the generators
next(result)
```

# Use Case
- For large numbers of items, generators reduce memory consumption and improve performance
- Very efficient for large datasets where not all values are needed simultaneously
- When all values are not needed at once, prefer generator expressions over list comprehensions

```python
# Solution 2
def gen_cube(values):
    ''' Calculates the cube of a list and returns the tuples in the list. '''
    for value in values:
        yield value, value ** 3

print(list(gen_cube(range(1,11))), end=' ')
```
