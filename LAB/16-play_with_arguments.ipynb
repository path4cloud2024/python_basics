## Default Paramaeter

```python
def multiply(num1=4, num2=2):
    """Calculating the mulitplication of to numbers"""
    return num1* num2
```

```python
multiply()  # without any parameter
```

```python
multiply(10) # with first parameter only, by default, it will consider this as first only
```

```python
multiply(15,3) # with both parameters
```

```python
multiply(num2=10) # explicitly defining second parameter
```

## Without Default Parameter

```python
def multiply(num1, num2):
    """Calculating the mulitplication of to numbers"""
    return num1* num2
```

```python
multiply() # if you don't provide any values, then it will TypeError
```

```python
multiply(15,5)
```

## Arbitrary Parameter

```python
def average(*args):
    """ calculating the average """
    return sum(args) / len(args)
```

```python
average(1, 4, 7, 9, 4)
```

```python
average(2, 5, 7)
```

## All types of Parameter

```python
def conversion(total,rate,*args):
    """ just craeting total of given total number with rate and given arguments"""
    return total * rate * sum(args)
```

```python
conversion(2, 4, 1, 2, 3, 4)
```

## Argument to Function: A Deeper Look

- Python Arguments are always passed by reference which is also called pass-by-object refernece
- It copies arguments object's reference, not the object itself because it can hit the memory slow execution (performance issues)
- We interact with an object via its reference. A referenec is the object's address in computer memory
- Check below examples for more understandings

```python
x = 7 # here we are assiging the value 7 to variable x so im memory x refres to object (address) which contans the value 7
```

```python
x
```

## Call the object reference with function `id`
- It will print the unique integer for that object
- No two separate objects can reside at the same address in memory

```python
id(x) # will print the unique idetifier of x in memory
```

```python
# Now, if we perform some operation with x, it will give new value based on operation but won't change the reference in memory
x * 2
```

```python
# let check the reference, it will be same
id(x)
```

```python
# still it won't change the reference
num = 5
num *= x
num
```

```python
# check the value of x and it's reference, it won't change
x
```

```python
id(x)
```
