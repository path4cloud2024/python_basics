# Arbitrary Arguments (*args and **kwargs) 

- Arbitrary Positional Arguments (*args): Allows a function to accept any number of non-keyword arguments. These are collected into a tuple.
- Arbitrary Keyword Arguments (**kwargs): Allows a function to accept any number of named (keyword) arguments. These are collected into a dictionary.

## *args (Arbitrary Positional Arguments)
- What it does
  - Allows you to pass any number of positional arguments
  - Stored as a tuple

```python
# function without positional arguments
def add_numbers(*args):                       # Here, it takes any number of arguments, can be called without any argument as well
    total = 0
    for num in args:
        total += num
    return total

print(add_numbers(1, 2, 3, 4))                # takes any number of arguments
#add_numbers(4)                               # works with single argument
#add_numbers()                                # works without any single argument since no positional arguments are defined
```

```python
# function with positional arguments and arbitrary arguments
def show_info(name, *args):                    # define one mandatory positional argument, and additional arguments are optional
    print("Name:", name)
    print("Other:", args)

show_info("Amit", 25, "DevOps", "Python")      # takes 'Amit' as name and '25, DevOps, Python' as additional argument
#show_info()                                     # since name is mandatory, if not given, then shows error "missing 1 required positional argument: 'name'"
#show_info("Amit")                               # here name is only mandatory, we can skip the additional arguments
```

## **kwargs (Arbitrary Keyword Arguments)
- What it does
  - Allows you to pass any number of keyword arguments
  - Stored as a dictionary
- If we want to pass the argument as key:value format without **kwargs, then we can specify with * at the beginning, and it forces the function not to take positional arguments, rather key:value format.
- We can pass * to force to take key:value format and the **kwargs, here, * takes only key, and that key must be defined while calling the function.
- if given with *args, then **kwargs must appear at the end of the function's parameter list after positional or *args arguments

```python
# function with * (forcing to take key:value argument
def mail(*, send, recipient):                                           # All parameters after * must be passed as keyword arguments
    print(f"From: {send} \nTo: {recipient}")

mail(send="amit@path4cloud.com", recipient="rahul@path4cloud.com")      # accepet only key:value format

# If we give in this format, then it will consider those as positional arguments and give an error
#mail("amit", "rahul")
```

```python
# function with * (forcing to take key:value argument, and then we can pass additional key:value with **kwargs

# All parameters after * must be passed as keyword arguments, **kwargs takes key:value
def mail(*, send, recipient, **kwargs):
    print(f"From: {send} \nTo: {recipient} \nsubject: {kwargs.get('Sub', 'Pass subject title as Sub=<tilte>')}") # to get the third argument from **kwargs, use .get()
mail(send="amit@path4cloud.com", recipient="rahul@path4cloud.com", Sub="test_email")                             # with additional key value argument        

# if no additional key value argument given, then it will print the default value from the function
#mail(send="amit@path4cloud.com", recipient="sam@path4cloud.com")
```

```python
# put a condition if you don't want to suggest the user pass an additional argument, or to pass a required default argument only

# All parameters after * must be passed as keyword arguments, **kwargs takes key:value
def mail(*, send, recipient, **kwargs):
    if kwargs:
        for key, value in kwargs.items():
            print(f"From: {send} \nTo: {recipient} \nsubject: {kwargs.get('Sub', 'Pass subject title as Sub=<tilte> as third arguments')}")
    else:
        print(f"From: {send} \nTo: {recipient}")
            
mail(send="amit@path4cloud.com", recipient="rahul@path4cloud.com", Sub="Test_Email")      

# if no additional key value argument given, then it will print the required arguments only, not **kwargs
#mail(send="amit@path4cloud.com", recipient="sam@path4cloud.com")      
```

```python
# We can have multiple **kwargs as well, or leave them empty as well. In the previous example, we tend to display only Sub if given
def info(*, name, age, **kwargs):
    if kwargs:
        print(f'name: {name} \nage: {age}')   # display the required arguments
        for key, value in kwargs.items():     # include if additional key:value argument given
            print(f'{key}:{value}')
    else:
        print(f'name: {name} \nage: {age}')

# passing only required arguments
#info(name="sam", age=34)

# passing multiple key:value arguments along with required arguments
info(name="sam", age=34, gender="male", role="DevOps")
```

```python
# function with ** kwargs only
def show_details(**kwargs):                 # here, **kwargs must be defined in key:value format if given, else empty kwargs are also accepted
    print(kwargs)

show_details(name="Amit", age=25)
#show_details()
#show_details('name', 'age')                 # if given as without key:value format, then it will consider those as positional arguments, since here, not defined, so shows error
```

```python
# function with ** kwargs and looping over the dictionary
def show_details(**kwargs):
    for key, value in kwargs.items():                      # **kwargs is stored as a dictionary so that we can iterate over it with the .items() method
        print(key, ":", value)

show_details(name="Amit", role="DevOps", skill="Python")
```

```python
# function with ** kwargs and conditional logic
def check_user(**kwargs):
    if 'admin' in kwargs:
        print("Admin user:", kwargs['admin'])
    else:
        print("Normal user")

check_user(admin="Amit")
check_user(user='Sam')
```

#### Using *args and **kwargs together

```python
def demo(a, *args, **kwargs):               # a is a positional parameter, then taking optional arguments, and then keyword arguments
    print("a:", a)
    print("args:", args)
    print("kwargs:", kwargs)

print("With positional args, *args and **kwargs")
demo(1, 2, 3, name="Amit", role="DevOps")
print("------------------")
print("With positional args and **kwargs")
demo(1, name="Amit", role="DevOps")
print("------------------")
print("With positional args only")
demo(1)
```

## Unpacking *args and **kwargs
- We can pass any iterable objects to a function with * (single parameter) and ** (with key:value format)

```python
# unpack *args
digits = [1, 2, 3, 4, 5]
def multiply(*args):
    total = 1
    for num in args:
        total *= num
    return total

print(f'with no arguments: {multiply()}')               # works without any argument since no positional argument is defined in the function
print(f'with single arguments: {multiply(2)}')          # works with a single argument as well
print(f'with multiple arguments: {multiply(2,10)}')     # works with multiple arguments as well

# calling the function with a list
print('----------------------')
print(f'passing list as *args and the result is: {multiply(*digits)}')  #here *args takes the element from the list and passes those as arguments
```

```python
# ** unpacking **kwargs
users = {'name': 'sam', 'age': 36, 'gender': 'male', 'role': 'DevOps'}  #make sure the dictionary has the required parameter key defined, order doesn't matter

#users = {'role': 'DevOps', 'name': 'sam', 'gender': 'male', 'age': 36}

# users = {'age': 36, 'name': 'sam'}  # will be accepted if you pass only required arguments

def info(*, name, age, **kwargs):
    if kwargs:
        print(f'name: {name} \nage: {age}')
        for key, value in kwargs.items():
            print(f'{key}:{value}')
    else:
        print(f'name: {name} \nage: {age}')

info(**users)
```

## When to use *args vs **kwargs

- use *args
  - Order matters
  - No need for names of arguments

```python
# DevOps Use Case: Running command on multiple servers
def run_command(cmd, *servers):
    for s in servers:
        print(f"{cmd} on {s}")

run_command("uptime", "server1", "server2")
```

- use **kwargs
  - Order doesn’t matter
  - You need named arguments

```python
# DevOps Use Case: Cloud resource creation
def create_instance(name, **config):
    print(f"Creating {name}")
    for k, v in config.items():
        print(f"{k}: {v}")

create_instance("web-server", cpu=2, ram="4GB", os="linux")
```

```python
# Real DevOps Use Case: Use Both
def deploy_service(service, *hosts, **settings):
    print(f"Deploying {service}")
    
    for h in hosts:
        print(f"→ Host: {h}")
    
    for k, v in settings.items():
        print(f"{k}: {v}")

deploy_service(
    "api",
    "server1", "server2",
    version="v2",
    env="prod",
    rollback=True
)
```
