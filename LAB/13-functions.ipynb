### function with multiple parameter

```python
def maximum(num1,num2,num3):
    """Calculating the maximum number from given list"""
    max_num = num1
    if num2 > max_num:
        max_num = num2
    if num3 > max_num:
        max_num = num3
    return max_num
```

```python
maximum(4,5,7)
```

```python
maximum('apple','orange','water melon') ## Yes thats true, < or > operation can be done against string as well, here first charater will be 
                                        ## checked in Alphatetic order and based on which is coming last, consider as maximum
```
