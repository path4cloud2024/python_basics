### function with multiple parameter

```python
def maximum(num1,num2,num3):
    """Calculating the maximum number from given values"""
    max_num = num1
    if num2 > max_num:
        max_num = num2
    if num3 > max_num:
        max_num = num3
    return max_num
```

```python
maximum(4,5,7)
```

# Python uses UNICODE - each character has a unicode integer value
- So, it does the lexicographical comparison
  - Strings are compared using their characters' numerical values
  - Uppercase letters 'A-Z' have values '65-90'
  - Lowercase letters 'a-z' have values '97-122'
  - So, lower case a is greater than upper case A.

```python
maximum('apple', 'orange', 'water melon') ## Yes, that's true, < or > operation can be done against a string as well, here the first character will be 
                                        ## checked in Alphatetic order and based on which comes last, consider as maximum
```
