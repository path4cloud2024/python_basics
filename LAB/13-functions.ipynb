```python
# Solution
def cube(number):
    ''' calculating the cube of give number'''
    return number ** 3
cube(3)
```

```python
def maximum(num1,num2,num3):
    """Calculating the maximum number from given values"""
    max_num = num1
    if num2 > max_num:
        max_num = num2
    if num3 > max_num:
        max_num = num3
    return max_num
```

```python
maximum(4,5,7)
```

# Lambda Function
- A lambda function is a small anonymous (unnamed) function
- It is written in one line
- Syntax
  - lambda arguments: expression

```python
# Example
add = lambda a, b: a + b
print(add(2, 3))
```

## Use case
- In automation Script, where we need to deal with list to perform some functions e.g. Convert servername from lower to upper using map function
- Filtering the data
- Sorting the data

```python
# Examples
nums = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, nums))
print(result)
```

```python
# Example
servers = ["web1", "web2", "web3"]

upper_servers = list(map(lambda s: s.upper(), servers))
print(upper_servers)
```

# Recursion Function
- Recursion is when a function calls itself to solve a problem.
- A function that calls itself until a condition is met.
- A function that calls itself until a condition is met.
- `Key Points`
  - Function calls itself
  - Must have base case
  - Used for repetitive problems
- `Common Mistakes`
  - No base case → infinite loop
  - Don't use too deep recursion

```python
# find the factorial of number 5
def factorial(n):
    if n == 1:   # base case
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
# find the factorial with loop
def factorial(n):
    result = 1
    
    for i in range(1, n + 1):
        result *= i   # result = result * i
    
    return result

print(factorial(5))
```

```python
def sum_of(n):
    if n == 0:
        return 0
    return n%10 + sum_of(n//10)

sum_of(7532)
```
