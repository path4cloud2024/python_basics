# File Handling

- File handling = working with files like:
  - reading logs
  - writing configs
  - processing data
- It has below functions:
  - Open ( ) function: used to open the file
    - syntax: open(filename, mode)
    - file mode:
      - `r`: open a file for reading (default mode)
      - `w`: to write into a file
      - `a`: to append to the existing file
      - `x`: to create a new file
      - `b`: for binary file (can be used as `rb` - open a file reading only binary format
      - `t`: text file
  - close( ) function: to close the file
  - read( ) function: to read the file
  - readline( ) function: to read the line from a file

## Creating a file, writing, appending, reading

```python
# Creating a new file
f= open('zmy_file.txt', 'x')  # better to define the extension of file
f.close()
```

```python
# writing into an existing file
f=open('zmy_file.txt', 'w')
f.write("Hello from Path4Cloud!!!, Welcome to Python Class")
f.close()
```

```python
# To Append to an existing file
f=open('zmy_file.txt','a')
f.write('\n\nWe are learning file handling with Python.')
f.close()
```

```python
# To read a file
f=open('zmy_file.txt', 'r')
data = f.read()
print(data)
f.close()
```

```python
# reading some specific character from the first line
f=open('zmy_file.txt', 'r')
data = f.read(24) # reading only the first 24 characters from the first line, including white space
print(data)
print(f.read())  # read remaining characters from the first line if any, else read all other lines
f.close()
```

```python
# to read only first line
f=open('zmy_file.txt', 'r')
print(f.readline())
f.close()
```

```python
# read line by line 
f=open('zmy_file.txt', 'r')
print(f.readline())
print(f.readline())
print(f.readline())
f.close()
```

```python
# reading number of lines
f=open('zmy_file.txt', 'r')
print(f.readlines(), end='\n')
f.close()
```

- readlines() prints the list of lines as a list element and print in one line, but here \n proves that it reading the next line
- If we want to print line by line then use for loop for cleaner output

```python
f=open('zmy_file.txt', 'r')
for lines in f.readlines():
    print(lines, end='')
f.close()
```

```python
# since read mode is the default, we can read the entire file without read mode
f=open('zmy_file.txt')
for lines in f:
    print(lines, end='')
f.close()
```

```python
# if any extra empty line, then we can get rid of that with strip( )
f=open('zmy_file.txt')
for lines in f:
    if lines.strip():
        print(lines.strip(),end='\n')
        # print(line.strip()) if line.strip() else None #shorter method
f.close()
```

## Reading and wrting the existing file at same time with `r+` mode which will append the content not overwrite the exsiting content

```python
f=open('zmy_file.txt', 'r+')
print(f.read())
print(f.write('\nI am a new line post reading'))  # show the number of written character as well in output
print("----------After writing----------")
f.seek(0)   # Once read function is invoked, pointer moved to the end of ilne so nothing left to read, so without this , it will read nothing, seek(0) will reset the pointer
print(f.read())
f.close()
```

- `r+` mode have a catch here, if you wanted to append the content at the end with this mode then first read the file and then write, it will not overwrite the content, but if write will be first before read then it will overwrite the content in first line and then read the remaining line

```python
f=open('zmy_file.txt', 'r+')
print(f.write('Hey!!'))  # show the number of written character as well in output
print(f.read())
print("----------After writing----------")
f.seek(0)   # Once read function is invoked, pointer moved to the end of the line, so nothing is left to read, so without this, it will read nothing, seek(0) will reset the pointer
print(f.read())
f.close()
```

## with statement

- `with` can be used while opening a file, advantage of with statement is that it will take care of closing the file.
- it uses a Context Manager. It automatically handles opening and most importantly closing the file for you, even if your code crashes or hits an error in the middle.
- When the code block inside the with statement finishes, Python immediately calls .close() behind the scenes.

```python
with open('zfile.txt', 'w') as fa:
    fa.write('I am sample text, and file gets auto closed')
```

- Create a new two files and write some content and then append the content from both files to a new file

```python
# Solution
f1=open('zfile1.txt','w') # here w mode will create the file if not existing
f2=open('zfile2.txt','w')

# writing the content to both files
f1.write('This content is from zfile1')
f2.write('This content is from zfile2')

# post writing, closing the file
f1.close()
f2.close()

# now we need to read and store the content into a variable so that we can feed it to file3
r1=open('zfile1.txt', 'r')
r2=open('zfile2.txt', 'r')

# storing the content into 2 separate files and merging the content into a third variable
c1=r1.read()
c2=r2.read()
c3=c1+'\n'+c2

# creating the third file and writing into the files
f3=open('zfile3.txt','w')
f3.write(c3)

# printing the merged content from both files
print(c3)

# closing all files
r1.close()
r2.close()
f3.close()
```

- Issues in your current code
  - ❌ 1. Too many open() / close()
  - Easy to forget closing files
- Risk of resource leaks
  - ❌ 2. Repetitive code
  - Same pattern repeated multiple times
- ❌ 3. Not scalable
  - If you want 5 files → code becomes messy

```python
# Improved Version with auto close
# write initial files
with open('zfile1.txt', 'w') as f1, open('zfile2.txt', 'w') as f2:
    f1.write('This content is from zfile1')
    f2.write('This content is from zfile2')

# read and merge
with open('zfile1.txt') as r1, open('zfile2.txt') as r2:
    c1 = r1.read()
    c2 = r2.read()

c3 = f"{c1}\n{c2}"

# write to file3
with open('zfile3.txt', 'w') as f3:
    f3.write(c3)

print(c3)
```

```python
# Scalable Version
files = ['zfile1.txt', 'zfile2.txt']

for i, file in enumerate(files, start=1):
    with open(file, 'w') as f:   # 'w' creates file if not exists
        f.write(f"This is content from {file}\n")

contents = []

for file in files:
    with open(file) as f:
        contents.append(f.read())

merged = "\n".join(contents)

with open('zfile3.txt', 'w') as f:
    f.write(merged)

print(merged)
```

```python
# Devops Version with os modules
import os

files = ['zfile1.txt', 'zfile2.txt']

# create files safely
for file in files:
    if not os.path.exists(file):
        with open(file, 'w') as f:
            f.write(f"Content for {file}\n")

# merge
merged = "\n".join(open(f).read() for f in files)

with open('zfile3.txt', 'w') as f:
    f.write(merged)

print(merged)
```

- Feed this content to one file "Welcome to Python Training", then find out the number of characters, lines, words and white space in this file.
- In the same file, find how many vowels are there

```python
# Solution
# Step 1: Write content
with open("zdata.txt", "w") as f:
    f.write("Welcome to Python Training")

# Step 2: Read content
with open("zdata.txt", "r") as f:
    content = f.read()

# Counts
char_count = len(content)
word_count = len(content.split())
line_count = content.count("\n") + 1
space_count = sum(1 for c in content if c.isspace())

vowels = "aeiouAEIOU"
vowel_count = sum(1 for c in content if c in vowels)

# Output
print("Content:", content)
print("Characters:", char_count)
print("Words:", word_count)
print("Lines:", line_count)
print("White spaces:", space_count)
print("Vowels:", vowel_count)
```
